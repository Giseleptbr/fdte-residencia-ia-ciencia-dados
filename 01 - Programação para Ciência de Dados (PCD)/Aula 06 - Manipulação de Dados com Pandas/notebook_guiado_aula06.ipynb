# === PRÁTICA 1: GROUPBY BÁSICO ===

## Objetivo
Usar groupby() para agregar dados por categoria.

## Tarefa
Dado um DataFrame com colunas 'categoria' e 'valor',
agrupar por categoria e somar os valores.

## Retorno Esperado
Series com soma por categoria.

## Dicas
```python
# Agrupar e agregar
df.groupby('coluna')['valores'].sum()

# Ou
df.groupby('coluna').sum()
```

**Tempo estimado:** 5 minutos

````python
def agrupar_e_somar(df):
    """
    Agrupa por 'categoria' e soma 'valor'.

    Args:
        df (pd.DataFrame): DataFrame com 'categoria' e 'valor'

    Returns:
        pd.Series: Soma por categoria
    """
    # === SEU CÓDIGO AQUI ===

    # Agrupar e somar
    resultado = df.groupby('categoria')['valor'].sum()

    return resultado

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de vendas
df_vendas = pd.DataFrame({
    'categoria': ['Eletrônicos', 'Livros', 'Eletrônicos', 'Roupas', 'Livros', 'Eletrônicos'],
    'produto': ['Mouse', 'Python', 'Teclado', 'Camiseta', 'Java', 'Monitor'],
    'valor': [45.00, 80.00, 120.00, 35.00, 75.00, 890.00]
})

print("\n=== VENDAS POR PRODUTO ===")
print(df_vendas)

# Agrupar e somar
total_por_categoria = agrupar_e_somar(df_vendas)

print("\n=== TOTAL POR CATEGORIA ===")
print(total_por_categoria)

print(f"\nCategoria com maior faturamento: {total_por_categoria.idxmax()}")
print(f"Valor: R$ {total_por_categoria.max():.2f}")

# Verificar teste
grader.check("p1")
````

# === PRÁTICA 2: MÚLTIPLAS AGREGAÇÕES ===

## Objetivo
Calcular múltiplas estatísticas com groupby.

## Tarefa
Dado DataFrame com 'cat' e 'val',
agrupar por 'cat' e calcular soma, média e contagem.

## Retorno Esperado
DataFrame com 3 colunas: sum, mean, count.

## Dicas
```python
# Múltiplas agregações
df.groupby('col')['val'].agg(['sum', 'mean', 'count'])

# Ou com dicionário
df.groupby('col').agg({'val': ['sum', 'mean', 'count']})
```

**Tempo estimado:** 6 minutos

````python
def calcular_multiplas_agregacoes(df):
    """
    Agrupa por 'cat' e calcula sum, mean, count de 'val'.

    Args:
        df (pd.DataFrame): DataFrame com 'cat' e 'val'

    Returns:
        pd.DataFrame: Agregações múltiplas
    """
    # === SEU CÓDIGO AQUI ===

    # Múltiplas agregações
    resultado = df.groupby('cat')['val'].agg(['sum', 'mean', 'count'])

    return resultado

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de exemplo
df_exemplo = pd.DataFrame({
    'cat': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'val': [10, 20, 15, 30, 25, 20, 35, 30]
})

print("\n=== DADOS ORIGINAIS ===")
print(df_exemplo)

# Calcular agregações
agregacoes = calcular_multiplas_agregacoes(df_exemplo)

print("\n=== AGREGAÇÕES POR CATEGORIA ===")
print(agregacoes)

print("\n=== ANÁLISE ===")
print(f"Categoria com maior soma: {agregacoes['sum'].idxmax()}")
print(f"Categoria com maior média: {agregacoes['mean'].idxmax()}")
print(f"Categoria mais frequente: {agregacoes['count'].idxmax()}")

# Verificar teste
grader.check("p2")
````

# === PRÁTICA 3: MERGE SIMPLES ===

## Objetivo
Combinar dois DataFrames usando merge.

## Tarefa
Dados dois DataFrames com coluna 'id' em comum,
fazer merge para combinar as informações.

## Retorno Esperado
DataFrame combinado com colunas de ambos.

## Dicas
```python
# Merge em coluna comum
pd.merge(df1, df2, on='coluna_comum')

# Ou método
df1.merge(df2, on='coluna_comum')
```

**Tempo estimado:** 5 minutos

````python
def fazer_merge_simples(df1, df2):
    """
    Faz merge de df1 e df2 na coluna 'id'.

    Args:
        df1 (pd.DataFrame): Primeiro DataFrame
        df2 (pd.DataFrame): Segundo DataFrame

    Returns:
        pd.DataFrame: DataFrames combinados
    """
    # === SEU CÓDIGO AQUI ===

    # Fazer merge
    resultado = pd.merge(df1, df2, on='id')

    return resultado

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrames de exemplo
df_clientes = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'cidade': ['SP', 'RJ', 'MG', 'SP']
})

df_pedidos = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'total_pedidos': [5, 3, 8, 2],
    'total_gasto': [450.00, 230.00, 890.00, 150.00]
})

print("\n=== CLIENTES ===")
print(df_clientes)

print("\n=== PEDIDOS ===")
print(df_pedidos)

# Fazer merge
df_completo = fazer_merge_simples(df_clientes, df_pedidos)

print("\n=== DADOS COMBINADOS ===")
print(df_completo)

print("\n=== ANÁLISE ===")
print(f"Cliente que mais gastou: {df_completo.loc[df_completo['total_gasto'].idxmax(), 'nome']}")
print(f"Cliente com mais pedidos: {df_completo.loc[df_completo['total_pedidos'].idxmax(), 'nome']}")

# Verificar teste
grader.check("p3")
````

# === PRÁTICA 4: PIVOT TABLE ===

## Objetivo
Criar pivot table para reorganizar dados.

## Tarefa
Dado DataFrame com 'produto', 'mes' e 'vendas',
criar pivot table com produtos nas linhas e meses nas colunas.

## Retorno Esperado
DataFrame pivotado (matriz produto × mês).

## Dicas
```python
# Pivot table
df.pivot_table(
    values='valores',
    index='linhas',
    columns='colunas',
    aggfunc='sum'
)
```

**Tempo estimado:** 7 minutos

````python
def criar_pivot_table(df):
    """
    Cria pivot table de vendas por produto e mês.

    Args:
        df (pd.DataFrame): DataFrame com 'produto', 'mes', 'vendas'

    Returns:
        pd.DataFrame: Pivot table
    """
    # === SEU CÓDIGO AQUI ===

    # Criar pivot table
    pivot = df.pivot_table(
        values='vendas',
        index='produto',
        columns='mes',
        aggfunc='sum'
    )

    return pivot

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de vendas mensais
df_vendas_mensais = pd.DataFrame({
    'produto': ['Mouse', 'Teclado', 'Monitor', 'Mouse', 'Teclado', 'Monitor'],
    'mes': ['Janeiro', 'Janeiro', 'Janeiro', 'Fevereiro', 'Fevereiro', 'Fevereiro'],
    'vendas': [100, 80, 30, 120, 90, 35]
})

print("\n=== DADOS ORIGINAIS ===")
print(df_vendas_mensais)

# Criar pivot table
pivot = criar_pivot_table(df_vendas_mensais)

print("\n=== PIVOT TABLE (Produto × Mês) ===")
print(pivot)

print("\n=== ANÁLISE ===")
print(f"Total por produto:\n{pivot.sum(axis=1)}")
print(f"\nTotal por mês:\n{pivot.sum(axis=0)}")

# Verificar teste
grader.check("p4")
````

# === PRÁTICA 5: TRANSFORMAR COLUNAS ===

## Objetivo
Criar nova coluna com cálculo baseado em outra.

## Tarefa
Dado DataFrame com coluna 'preco',
adicionar coluna 'preco_com_desconto' aplicando 10% de desconto.

## Retorno Esperado
DataFrame com nova coluna.

## Dicas
```python
# Criar nova coluna
df['nova_coluna'] = df['coluna'] * 0.9
```

**Tempo estimado:** 4 minutos

````python
def adicionar_coluna_calculada(df):
    """
    Adiciona coluna com 10% de desconto.

    Args:
        df (pd.DataFrame): DataFrame com 'preco'

    Returns:
        pd.DataFrame: DataFrame com nova coluna
    """
    # === SEU CÓDIGO AQUI ===

    # Criar cópia para não modificar original
    df_novo = df.copy()

    # Adicionar coluna com desconto
    df_novo['preco_com_desconto'] = df_novo['preco'] * 0.9

    return df_novo

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de produtos
df_produtos = pd.DataFrame({
    'produto': ['Mouse', 'Teclado', 'Monitor', 'Webcam'],
    'preco': [50.00, 120.00, 900.00, 250.00]
})

print("\n=== PREÇOS ORIGINAIS ===")
print(df_produtos)

# Adicionar coluna de desconto
df_com_desconto = adicionar_coluna_calculada(df_produtos)

print("\n=== COM DESCONTO DE 10% ===")
print(df_com_desconto)

# Calcular economia
df_com_desconto['economia'] = df_com_desconto['preco'] - df_com_desconto['preco_com_desconto']
print("\n=== ECONOMIA POR PRODUTO ===")
print(df_com_desconto[['produto', 'economia']])

# Verificar teste
grader.check("p5")
````

# === PRÁTICA 6: APPLY COM FUNÇÃO CUSTOMIZADA ===

## Objetivo
Usar apply() para transformar dados com função customizada.

## Tarefa
Dado DataFrame com 'valor', criar coluna 'classificacao':
- 'Alto' se valor >= 50
- 'Médio' se 20 <= valor < 50
- 'Baixo' se valor < 20

## Retorno Esperado
DataFrame com coluna 'classificacao'.

## Dicas
```python
# Definir função
def classificar(valor):
    if valor >= 50:
        return 'Alto'
    # ...

# Aplicar
df['nova'] = df['coluna'].apply(classificar)
```

**Tempo estimado:** 7 minutos

````python
def aplicar_funcao_customizada(df):
    """
    Classifica valores em Alto/Médio/Baixo.

    Args:
        df (pd.DataFrame): DataFrame com 'valor'

    Returns:
        pd.DataFrame: DataFrame com 'classificacao'
    """
    # === SEU CÓDIGO AQUI ===

    # Criar cópia
    df_novo = df.copy()

    # Definir função de classificação
    def classificar(valor):
        if valor >= 50:
            return 'Alto'
        elif valor >= 20:
            return 'Médio'
        else:
            return 'Baixo'

    # Aplicar função
    df_novo['classificacao'] = df_novo['valor'].apply(classificar)

    return df_novo

    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de exemplo
df_valores = pd.DataFrame({
    'item': ['A', 'B', 'C', 'D', 'E', 'F'],
    'valor': [10, 25, 55, 15, 45, 70]
})

print("\n=== DADOS ORIGINAIS ===")
print(df_valores)

# Aplicar classificação
df_classificado = aplicar_funcao_customizada(df_valores)

print("\n=== COM CLASSIFICAÇÃO ===")
print(df_classificado)

# Contar por classificação
print("\n=== CONTAGEM POR CLASSIFICAÇÃO ===")
print(df_classificado['classificacao'].value_counts())

# Verificar teste
grader.check("p6")
````
