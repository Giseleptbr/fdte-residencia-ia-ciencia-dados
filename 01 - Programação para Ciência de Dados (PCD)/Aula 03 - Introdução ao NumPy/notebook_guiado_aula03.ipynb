# === PR√ÅTICA 2: INDEXA√á√ÉO E SLICING ===

## Objetivo
Extrair dados usando indexa√ß√£o multidimensional.

## Tarefa
Extrair a **segunda coluna** de uma matriz 2D.

## Retorno Esperado
Array 1D com todos os valores da coluna √≠ndice 1.

## Dicas
```python
# Acessar linha
matriz[0]      # Primeira linha
matriz[0, :]   # Primeira linha (expl√≠cito)

# Acessar coluna
matriz[:, 0]   # Primeira coluna
matriz[:, 1]   # Segunda coluna

# Elemento espec√≠fico
matriz[linha, coluna]
```

````python
def extrair_segunda_coluna(matriz):
    """
    Extrai a segunda coluna (√≠ndice 1) de uma matriz.

    Args:
        matriz (np.ndarray): Matriz 2D

    Returns:
        np.ndarray: Array 1D com valores da segunda coluna
    """
    # === SEU C√ìDIGO AQUI ===

    # Extrair coluna de √≠ndice 1
    coluna = matriz[:, 1]

    return coluna

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar matriz de exemplo
matriz_vendas = np.array([
    [100, 150, 200],  # Produto A
    [80, 120, 160],   # Produto B
    [90, 110, 140]    # Produto C
])

print("\n=== MATRIZ DE VENDAS (3 produtos x 3 meses) ===")
print(matriz_vendas)
print("\nColunas: Jan, Fev, Mar")

# Extrair segunda coluna (Fevereiro)
vendas_fev = extrair_segunda_coluna(matriz_vendas)

print("\n=== VENDAS DE FEVEREIRO ===")
print(f"Produto A: {vendas_fev[0]}")
print(f"Produto B: {vendas_fev[1]}")
print(f"Produto C: {vendas_fev[2]}")
print(f"\nTotal: {vendas_fev.sum()}")

# Verificar teste
grader.check("p2")
````

# === PR√ÅTICA 3: BOOLEAN INDEXING ===

## Objetivo
Filtrar array usando condi√ß√µes booleanas.

## Tarefa
Filtrar apenas valores **maiores** que um limiar especificado.

## Retorno Esperado
Array com apenas os valores que atendem √† condi√ß√£o.

## Dicas
```python
# Criar m√°scara booleana
mask = arr > 5

# Aplicar m√°scara
resultado = arr[mask]

# Forma compacta
resultado = arr[arr > 5]
```

````python
def filtrar_maiores(arr, limiar):
    """
    Filtra valores maiores que o limiar.

    Args:
        arr (np.ndarray): Array de entrada
        limiar (float): Valor de corte

    Returns:
        np.ndarray: Array com valores > limiar
    """
    # === SEU C√ìDIGO AQUI ===

    # Aplicar filtro booleano
    resultado = arr[arr > limiar]

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar dados de temperatura
np.random.seed(42)
temperaturas = np.random.uniform(15, 35, 20)

print("\n=== TEMPERATURAS (20 medi√ß√µes) ===")
print(f"Todas: {temperaturas}")
print(f"\nM√©dia: {temperaturas.mean():.1f}¬∞C")

# Filtrar temperaturas altas (> 28¬∞C)
temp_altas = filtrar_maiores(temperaturas, 28)

print("\n=== TEMPERATURAS ALTAS (> 28¬∞C) ===")
print(f"Valores: {temp_altas}")
print(f"Quantidade: {len(temp_altas)} medi√ß√µes")
print(f"Percentual: {len(temp_altas)/len(temperaturas)*100:.1f}%")

# Verificar teste
grader.check("p3")
````

# === PR√ÅTICA 4: BROADCASTING ===

## Objetivo
Realizar opera√ß√µes entre arrays de shapes diferentes.

## Tarefa
Somar uma matriz 2D com um vetor 1D usando broadcasting.

## Retorno Esperado
Matriz onde cada linha foi somada com o vetor.

## Dicas
```python
# Broadcasting autom√°tico
matriz = np.array([[1, 2], [3, 4]])  # (2, 2)
vetor = np.array([10, 20])            # (2,)

# Soma com broadcasting
resultado = matriz + vetor
# [[11, 22],
#  [13, 24]]
```

````python
def somar_com_broadcasting(matriz, vetor):
    """
    Soma matriz 2D com vetor 1D usando broadcasting.

    Args:
        matriz (np.ndarray): Matriz 2D
        vetor (np.ndarray): Vetor 1D

    Returns:
        np.ndarray: Resultado da soma
    """
    # === SEU C√ìDIGO AQUI ===

    # Broadcasting autom√°tico
    resultado = matriz + vetor

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 6: OPERA√á√ïES COM AXIS ===

## Objetivo
Aplicar agrega√ß√µes em dimens√µes espec√≠ficas.

## Tarefa
Calcular a soma de cada **linha** de uma matriz (axis=1).

## Retorno Esperado
Array 1D com a soma de cada linha.

## Dicas
```python
# Sem axis: opera em tudo
np.sum(matriz)  # Soma total

# axis=0: ao longo das linhas (colapsa linhas)
np.sum(matriz, axis=0)  # Soma de cada coluna

# axis=1: ao longo das colunas (colapsa colunas)
np.sum(matriz, axis=1)  # Soma de cada linha
```

````python
def somar_linhas(matriz):
    """
    Calcula a soma de cada linha.

    Args:
        matriz (np.ndarray): Matriz 2D

    Returns:
        np.ndarray: Array com soma de cada linha
    """
    # === SEU C√ìDIGO AQUI ===

    # Somar ao longo do axis=1 (colunas)
    somas = np.sum(matriz, axis=1)

    return somas

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Matriz de vendas trimestrais
vendas_trimestre = np.array([
    [100, 150, 200, 180],  # Produto A
    [80, 120, 160, 140],   # Produto B
    [90, 110, 140, 130]    # Produto C
])

print("\n=== VENDAS POR TRIMESTRE ===")
print("Produtos (linhas) x Meses (colunas)")
print(vendas_trimestre)

# Calcular total por produto
total_por_produto = somar_linhas(vendas_trimestre)

print("\n=== TOTAL POR PRODUTO (4 meses) ===")
produtos = ['Produto A', 'Produto B', 'Produto C']
for i, produto in enumerate(produtos):
    print(f"{produto}: {total_por_produto[i]}")

# Demonstrar axis=0 tamb√©m
print("\n=== TOTAL POR M√äS (axis=0) ===")
total_por_mes = np.sum(vendas_trimestre, axis=0)
meses = ['M√™s 1', 'M√™s 2', 'M√™s 3', 'M√™s 4']
for i, mes in enumerate(meses):
    print(f"{mes}: {total_por_mes[i]}")

# Verificar teste
grader.check("p6")
````

# === PR√ÅTICA 7: np.where() ===

## Objetivo
Usar `np.where()` para sele√ß√£o condicional.

## Tarefa
Substituir valores maiores que um limiar por um valor, e menores por outro.

## Retorno Esperado
Array com valores substitu√≠dos conforme condi√ß√£o.

## Dicas
```python
# np.where(condi√ß√£o, valor_se_true, valor_se_false)

arr = np.array([1, 5, 3, 8, 2])

# Se > 5 ent√£o 'alto', sen√£o 'baixo'
resultado = np.where(arr >= 5, 'alto', 'baixo')

# Pode usar os pr√≥prios valores
resultado = np.where(arr >= 5, arr * 10, arr)
```

````python
def substituir_com_where(arr, limiar, valor_alto, valor_baixo):
    """
    Substitui valores usando np.where().

    Args:
        arr (np.ndarray): Array de entrada
        limiar (float): Valor de corte
        valor_alto: Valor para elementos > limiar
        valor_baixo: Valor para elementos <= limiar

    Returns:
        np.ndarray: Array com valores substitu√≠dos
    """
    # === SEU C√ìDIGO AQUI ===

    # Usar np.where para substituir
    resultado = np.where(arr >= limiar, valor_alto, valor_baixo)

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Dados de temperatura
np.random.seed(42)
temperaturas = np.random.uniform(15, 35, 10)

print("\n=== CLASSIFICA√á√ÉO DE TEMPERATURAS ===")
print(f"Originais: {temperaturas}")

# Classificar: > 25¬∞C = 'quente', <= 25¬∞C = 'frio'
classificacao = substituir_com_where(temperaturas, 25, 'quente', 'frio')

print("\n=== RESULTADO ===")
for i, (temp, classe) in enumerate(zip(temperaturas, classificacao)):
    print(f"Medi√ß√£o {i+1}: {temp:.1f}¬∞C ‚Üí {classe}")

# Contar quantos de cada
print(f"\nQuentes: {np.sum(classificacao == 'quente')}")
print(f"Frios: {np.sum(classificacao == 'frio')}")

# Verificar teste
grader.check("p7")
````

# === PR√ÅTICA 8: COMPARA√á√ÉO DE DESEMPENHO ===

## Objetivo
Comparar desempenho de loop vs vetoriza√ß√£o.

## Tarefa
Medir tempo de execu√ß√£o de:
1. Loop Python para elevar ao quadrado
2. Opera√ß√£o vetorizada NumPy

## Retorno Esperado
Dicion√°rio com tempos de execu√ß√£o.

## Dicas
```python
import time

# Medir tempo
start = time.time()
# ... c√≥digo ...
tempo = time.time() - start

# Loop
resultado = []
for x in arr:
    resultado.append(x ** 2)

# Vetorizado
resultado = arr ** 2
```

````python
def comparar_performance(arr):
    """
    Compara performance de loop vs vetoriza√ß√£o.

    Args:
        arr (np.ndarray): Array de entrada

    Returns:
        dict: Tempos de execu√ß√£o
    """
    # === SEU C√ìDIGO AQUI ===

    # M√©todo 1: Loop Python
    start = time.time()
    resultado_loop = []
    for x in arr:
        resultado_loop.append(x ** 2)
    tempo_loop = time.time() - start

    # M√©todo 2: Vetorizado
    start = time.time()
    resultado_vetorizado = arr ** 2
    tempo_vetorizado = time.time() - start

    return {
        'tempo_loop': tempo_loop,
        'tempo_vetorizado': tempo_vetorizado,
        'speedup': tempo_loop / tempo_vetorizado if tempo_vetorizado > 0 else 0
    }

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar com diferentes tamanhos

print("\n=== COMPARA√á√ÉO DE PERFORMANCE ===\n")

tamanhos = [1000, 10000, 100000]

for n in tamanhos:
    arr = np.arange(n)
    resultado = comparar_performance(arr)

    print(f"üìä Array com {n:,} elementos:")
    print(f"   Loop:       {resultado['tempo_loop']:.6f}s")
    print(f"   Vetorizado: {resultado['tempo_vetorizado']:.6f}s")
    print(f"   Speedup:    {resultado['speedup']:.1f}x mais r√°pido")
    print()

print("üí° NumPy √© MUITO mais r√°pido que loops Python!")
print("   Sempre prefira opera√ß√µes vetorizadas.")

# Verificar teste
grader.check("p8")
````
