# === PR√ÅTICA 1: CRIAR SERIES ===

## Objetivo
Criar uma Series Pandas a partir de uma lista.

## Tarefa
Criar uma Series com 5 notas de alunos: [7.5, 8.0, 6.5, 9.0, 7.0]

## Retorno Esperado
Series Pandas com os valores fornecidos.

## Dicas
```python
# Criar Series a partir de lista
s = pd.Series([1, 2, 3, 4, 5])

# Com √≠ndice customizado
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
```

**Tempo estimado:** 3 minutos

````python
def criar_series_notas():
    """
    Cria Series com notas de alunos.

    Returns:
        pd.Series: Series com 5 notas
    """
    # === SEU C√ìDIGO AQUI ===

    # Criar lista de notas
    notas = [7.5, 8.0, 6.5, 9.0, 7.0]

    # Criar Series
    s = pd.Series(notas)

    return s

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
notas = criar_series_notas()

print("\n=== SERIES DE NOTAS ===")
print(notas)
print(f"\nTipo: {type(notas)}")
print(f"Tamanho: {len(notas)}")
print(f"Dtype: {notas.dtype}")
print(f"\nM√©dia: {notas.mean():.2f}")
print(f"Maior nota: {notas.max():.2f}")
print(f"Menor nota: {notas.min():.2f}")

# Verificar teste
grader.check("p1")
````

# === PR√ÅTICA 2: CRIAR DATAFRAME ===

## Objetivo
Criar um DataFrame a partir de um dicion√°rio.

## Tarefa
Criar DataFrame com informa√ß√µes de 4 alunos:
- Nome: ['Ana', 'Bruno', 'Carlos', 'Diana']
- Idade: [20, 22, 21, 23]
- Nota: [8.5, 7.0, 9.0, 8.0]

## Retorno Esperado
DataFrame 4x3 (4 linhas, 3 colunas).

## Dicas
```python
# Criar DataFrame de dicion√°rio
df = pd.DataFrame({
    'coluna1': [1, 2, 3],
    'coluna2': ['a', 'b', 'c']
})
```

**Tempo estimado:** 5 minutos

````python
def criar_dataframe_alunos():
    """
    Cria DataFrame com dados de alunos.

    Returns:
        pd.DataFrame: DataFrame com nome, idade e nota
    """
    # === SEU C√ìDIGO AQUI ===

    # Criar dicion√°rio com dados
    dados = {
        'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
        'Idade': [20, 22, 21, 23],
        'Nota': [8.5, 7.0, 9.0, 8.0]
    }

    # Criar DataFrame
    df = pd.DataFrame(dados)

    return df

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
df_alunos = criar_dataframe_alunos()

print("\n=== DATAFRAME DE ALUNOS ===")
print(df_alunos)
print(f"\nShape: {df_alunos.shape}")
print(f"Colunas: {list(df_alunos.columns)}")
print(f"\nInfo:")
df_alunos.info()
print(f"\nEstat√≠sticas:")
print(df_alunos.describe())

# Verificar teste
grader.check("p2")
````

# === PR√ÅTICA 3: SELECIONAR COLUNA ===

## Objetivo
Selecionar uma coluna espec√≠fica de um DataFrame.

## Tarefa
Dado um DataFrame com colunas 'Nome', 'Idade', 'Nota',
selecionar apenas a coluna 'Nome'.

## Retorno Esperado
Series com os nomes.

## Dicas
```python
# Selecionar uma coluna (retorna Series)
df['coluna']
df.coluna  # Se o nome n√£o tem espa√ßos

# Selecionar m√∫ltiplas colunas (retorna DataFrame)
df[['coluna1', 'coluna2']]
```

**Tempo estimado:** 3 minutos

````python
def selecionar_coluna_nome(df):
    """
    Seleciona a coluna 'Nome' do DataFrame.

    Args:
        df (pd.DataFrame): DataFrame com coluna 'Nome'

    Returns:
        pd.Series: Coluna com nomes
    """
    # === SEU C√ìDIGO AQUI ===

    # Selecionar coluna Nome
    nomes = df['Nome']

    return nomes

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Usar DataFrame da pr√°tica anterior
df_alunos = criar_dataframe_alunos()

print("\n=== DATAFRAME ORIGINAL ===")
print(df_alunos)

# Selecionar coluna Nome
nomes = selecionar_coluna_nome(df_alunos)

print("\n=== COLUNA 'NOME' ===")
print(nomes)
print(f"\nTipo: {type(nomes)}")
print(f"Tamanho: {len(nomes)}")

# Verificar teste
grader.check("p3")
````

# === PR√ÅTICA 4: USAR loc[] ===

## Objetivo
Usar `loc[]` para sele√ß√£o baseada em r√≥tulos.

## Tarefa
Dado um DataFrame, selecionar a primeira linha (√≠ndice 0)
usando `loc[]`.

## Retorno Esperado
Series com os dados da primeira linha.

## Dicas
```python
# loc[linha, coluna]
df.loc[0]              # Primeira linha
df.loc[0, 'coluna']    # C√©lula espec√≠fica
df.loc[0:2]            # Linhas 0, 1, 2 (inclusivo!)
df.loc[:, 'coluna']    # Todas as linhas, uma coluna
```

**Tempo estimado:** 4 minutos

````python
def selecionar_com_loc(df):
    """
    Seleciona primeira linha usando loc[].

    Args:
        df (pd.DataFrame): DataFrame de entrada

    Returns:
        pd.Series: Primeira linha
    """
    # === SEU C√ìDIGO AQUI ===

    # Selecionar primeira linha com loc
    primeira_linha = df.loc[0]

    return primeira_linha

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Usar DataFrame de alunos
df_alunos = criar_dataframe_alunos()

print("\n=== DATAFRAME COMPLETO ===")
print(df_alunos)

# Selecionar primeira linha
primeira_linha = selecionar_com_loc(df_alunos)

print("\n=== PRIMEIRA LINHA (loc[0]) ===")
print(primeira_linha)
print(f"\nNome: {primeira_linha['Nome']}")
print(f"Idade: {primeira_linha['Idade']}")
print(f"Nota: {primeira_linha['Nota']}")

# Demonstrar outras sele√ß√µes
print("\n=== OUTRAS SELE√á√ïES COM loc[] ===")
print("\nLinhas 0 a 2:")
print(df_alunos.loc[0:2])

print("\nC√©lula espec√≠fica (linha 0, coluna 'Nome'):")
print(df_alunos.loc[0, 'Nome'])

# Verificar teste
grader.check("p4")
````

# === PR√ÅTICA 5: FILTRAGEM (BOOLEAN INDEXING) ===

## Objetivo
Filtrar linhas usando condi√ß√µes booleanas.

## Tarefa
Dado um DataFrame com coluna 'idade',
filtrar apenas pessoas com idade >= 18.

## Retorno Esperado
DataFrame com apenas maiores de idade.

## Dicas
```python
# Criar m√°scara booleana
mask = df['coluna'] > valor

# Aplicar m√°scara
resultado = df[mask]

# Forma compacta
resultado = df[df['coluna'] > valor]
```

**Tempo estimado:** 5 minutos

````python
def filtrar_maiores_idade(df):
    """
    Filtra pessoas com idade >= 18.

    Args:
        df (pd.DataFrame): DataFrame com coluna 'idade'

    Returns:
        pd.DataFrame: Apenas maiores de idade
    """
    # === SEU C√ìDIGO AQUI ===

    # Filtrar com boolean indexing
    maiores = df[df['idade'] >= 18]

    return maiores

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de exemplo
df_pessoas = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'idade': [17, 22, 15, 19, 16],
    'cidade': ['SP', 'RJ', 'SP', 'MG', 'RJ']
})

print("\n=== DATAFRAME ORIGINAL ===")
print(df_pessoas)

# Filtrar maiores de idade
maiores = filtrar_maiores_idade(df_pessoas)

print("\n=== MAIORES DE IDADE (>= 18) ===")
print(maiores)
print(f"\nTotal de maiores: {len(maiores)} de {len(df_pessoas)}")

# Verificar teste
grader.check("p5")
````

# === PR√ÅTICA 6: ORDENA√á√ÉO ===

## Objetivo
Ordenar DataFrame por uma coluna.

## Tarefa
Dado um DataFrame com coluna 'valor',
ordenar por essa coluna em ordem crescente.

## Retorno Esperado
DataFrame ordenado.

## Dicas
```python
# Ordenar por uma coluna
df.sort_values('coluna')

# Ordem decrescente
df.sort_values('coluna', ascending=False)

# M√∫ltiplas colunas
df.sort_values(['col1', 'col2'])
```

**Tempo estimado:** 4 minutos

````python
def ordenar_por_coluna(df):
    """
    Ordena DataFrame por coluna 'valor'.

    Args:
        df (pd.DataFrame): DataFrame com coluna 'valor'

    Returns:
        pd.DataFrame: DataFrame ordenado
    """
    # === SEU C√ìDIGO AQUI ===

    # Ordenar por valor (crescente)
    df_ordenado = df.sort_values('valor')

    return df_ordenado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de vendas
df_vendas = pd.DataFrame({
    'produto': ['Mouse', 'Teclado', 'Monitor', 'Webcam', 'Headset'],
    'valor': [45.90, 120.00, 890.00, 250.00, 180.00],
    'estoque': [50, 30, 15, 25, 40]
})

print("\n=== VENDAS (ORIGINAL) ===")
print(df_vendas)

# Ordenar por valor
df_ordenado = ordenar_por_coluna(df_vendas)

print("\n=== VENDAS ORDENADAS POR VALOR (CRESCENTE) ===")
print(df_ordenado)

# Demonstrar ordem decrescente
print("\n=== VENDAS ORDENADAS POR VALOR (DECRESCENTE) ===")
print(df_vendas.sort_values('valor', ascending=False))

# Verificar teste
grader.check("p6")
````

# === PR√ÅTICA 7: ESTAT√çSTICAS B√ÅSICAS ===

## Objetivo
Calcular estat√≠sticas de uma coluna.

## Tarefa
Dado um DataFrame com coluna 'notas',
calcular a m√©dia usando `.mean()`.

## Retorno Esperado
Valor float com a m√©dia.

## Dicas
```python
# Estat√≠sticas de uma coluna
df['coluna'].mean()    # M√©dia
df['coluna'].median()  # Mediana
df['coluna'].std()     # Desvio padr√£o
df['coluna'].min()     # M√≠nimo
df['coluna'].max()     # M√°ximo
```

**Tempo estimado:** 3 minutos

````python
def calcular_media_coluna(df):
    """
    Calcula m√©dia da coluna 'notas'.

    Args:
        df (pd.DataFrame): DataFrame com coluna 'notas'

    Returns:
        float: M√©dia das notas
    """
    # === SEU C√ìDIGO AQUI ===

    # Calcular m√©dia
    media = df['notas'].mean()

    return media

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de notas
df_notas = pd.DataFrame({
    'aluno': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'notas': [8.5, 7.0, 9.5, 6.5, 8.0]
})

print("\n=== NOTAS DOS ALUNOS ===")
print(df_notas)

# Calcular m√©dia
media = calcular_media_coluna(df_notas)
print(f"\n=== ESTAT√çSTICAS ===")
print(f"M√©dia: {media:.2f}")

# Outras estat√≠sticas
print(f"Mediana: {df_notas['notas'].median():.2f}")
print(f"Desvio padr√£o: {df_notas['notas'].std():.2f}")
print(f"Nota m√°xima: {df_notas['notas'].max():.2f}")
print(f"Nota m√≠nima: {df_notas['notas'].min():.2f}")

# Verificar teste
grader.check("p7")
````

# === PR√ÅTICA 8: VALORES FALTANTES ===

## Objetivo
Identificar valores faltantes (NaN).

## Tarefa
Dado um DataFrame, contar quantos valores faltantes
existem em cada coluna.

## Retorno Esperado
Series com contagem de NaN por coluna.

## Dicas
```python
# Verificar valores faltantes
df.isnull()       # M√°scara booleana
df.isnull().sum() # Contar por coluna

# Percentual de faltantes
(df.isnull().sum() / len(df)) * 100
```

**Tempo estimado:** 4 minutos

````python
def contar_valores_faltantes(df):
    """
    Conta valores faltantes por coluna.

    Args:
        df (pd.DataFrame): DataFrame de entrada

    Returns:
        pd.Series: Contagem de NaN por coluna
    """
    # === SEU C√ìDIGO AQUI ===

    # Contar valores faltantes
    contagem = df.isnull().sum()

    return contagem

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame com valores faltantes
df_com_nan = pd.DataFrame({
    'nome': ['Ana', 'Bruno', np.nan, 'Diana', 'Eduardo'],
    'idade': [25, np.nan, 30, 28, np.nan],
    'cidade': ['SP', 'RJ', 'MG', np.nan, 'SP']
})

print("\n=== DATAFRAME COM VALORES FALTANTES ===")
print(df_com_nan)

# Contar valores faltantes
contagem = contar_valores_faltantes(df_com_nan)

print("\n=== CONTAGEM DE VALORES FALTANTES ===")
print(contagem)

# Percentual
print("\n=== PERCENTUAL DE VALORES FALTANTES ===")
percentual = (contagem / len(df_com_nan)) * 100
print(percentual)

# Verificar teste
grader.check("p8")
````
