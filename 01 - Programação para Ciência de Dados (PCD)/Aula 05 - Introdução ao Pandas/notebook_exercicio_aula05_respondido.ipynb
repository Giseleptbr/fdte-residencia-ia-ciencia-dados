# Exerc√≠cio 1: Series com √çndice Customizado

## Descri√ß√£o
Crie uma Series com temperaturas da semana, usando os dias da semana como √≠ndice.

Temperaturas:
- Segunda: 25¬∞C
- Ter√ßa: 27¬∞C
- Quarta: 23¬∞C
- Quinta: 26¬∞C
- Sexta: 28¬∞C
- S√°bado: 30¬∞C
- Domingo: 29¬∞C

## Exemplo
```python
criar_series_temperaturas()
# Segunda    25.0
# Ter√ßa      27.0
# ...
```

## Dicas
- Use `pd.Series(dados, index=labels)`
- Os dados podem ser uma lista ou dicion√°rio

**Tempo estimado:** 7 minutos

````python
def criar_series_temperaturas():
    """
    Cria Series de temperaturas com dias da semana como √≠ndice.

    Returns:
        pd.Series: Temperaturas por dia da semana
    """
    # === SEU C√ìDIGO AQUI ===
    temperaturas = [25, 27, 23, 26, 28, 30, 29]
    dias = ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta', 'S√°bado', 'Domingo']
    return pd.Series(temperaturas, index=dias)


    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q1 ===")
temperaturas = criar_series_temperaturas()
print(temperaturas)

print(f"\nTemperatura m√©dia da semana: {temperaturas.mean():.1f}¬∞C")
print(f"Dia mais quente: {temperaturas.idxmax()} ({temperaturas.max():.1f}¬∞C)")
print(f"Dia mais frio: {temperaturas.idxmin()} ({temperaturas.min():.1f}¬∞C)")

# Acessar valores
print(f"\nTemperatura na Sexta: {temperaturas['Sexta']:.1f}¬∞C")
print(f"Fim de semana: \n{temperaturas[['S√°bado', 'Domingo']]}")

grader.check("q1")
````

# Exerc√≠cio 2: DataFrame com C√°lculos

## Descri√ß√£o
Crie um DataFrame de vendas com:
- Produtos: ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Webcam']
- Pre√ßos: [3500.00, 45.00, 120.00, 890.00, 250.00]
- Quantidades vendidas: [5, 20, 15, 8, 12]

Adicione uma coluna 'total' calculada como: pre√ßo √ó quantidade

## Exemplo
```python
criar_dataframe_vendas()
#      produto   preco  quantidade    total
# 0   Notebook  3500.0           5  17500.0
# 1      Mouse    45.0          20    900.0
# ...
```

## Dicas
- Crie DataFrame com 3 colunas primeiro
- Adicione coluna 'total' com: `df['total'] = df['preco'] * df['quantidade']`

**Tempo estimado:** 8 minutos

````python
def criar_dataframe_vendas():
    """
    Cria DataFrame de vendas com coluna total calculada.

    Returns:
        pd.DataFrame: DataFrame com vendas e total
    """
    # === SEU C√ìDIGO AQUI ===
    produtos = ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Webcam']
    precos = [3500.00, 45.00, 120.00, 890.00, 250.00]
    quantidades = [5, 20, 15, 8, 12]

    df = pd.DataFrame({
        'produto': produtos,
        'preco': precos,
        'quantidade': quantidades
    })
    df['total'] = df['preco'] * df['quantidade']
    return df

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q2 ===")
vendas = criar_dataframe_vendas()
print(vendas)

print(f"\nShape: {vendas.shape}")
print(f"Colunas: {list(vendas.columns)}")

print("\n=== AN√ÅLISE DE VENDAS ===")
print(f"Total de vendas: R$ {vendas['total'].sum():,.2f}")
print(f"Produto com maior faturamento: {vendas.loc[vendas['total'].idxmax(), 'produto']}")
print(f"Valor: R$ {vendas['total'].max():,.2f}")

# Ordenar por total
print("\n=== TOP 3 PRODUTOS (por faturamento) ===")
print(vendas.sort_values('total', ascending=False).head(3)[['produto', 'total']])

grader.check("q2")
````

# Exerc√≠cio 3: Filtragem com M√∫ltiplas Condi√ß√µes

## Descri√ß√£o
Dado um DataFrame com colunas 'preco' e 'estoque',
filtre produtos que atendem AMBAS as condi√ß√µes:
- Pre√ßo >= 100
- Estoque > 20

Esses s√£o considerados "produtos premium em estoque".

## Exemplo
```python
df = pd.DataFrame({
    'produto': ['A', 'B', 'C'],
    'preco': [50, 150, 200],
    'estoque': [10, 25, 30]
})
filtrar_produtos_premium(df)
#   produto  preco  estoque
# 1       B    150       25
# 2       C    200       30
```

## Dicas
- Use `&` (AND) entre condi√ß√µes
- Sempre use par√™nteses: `(cond1) & (cond2)`

**Tempo estimado:** 8 minutos

````python
def filtrar_produtos_premium(df):
    """
    Filtra produtos com pre√ßo >= 100 e estoque > 20.

    Args:
        df (pd.DataFrame): DataFrame com colunas 'preco' e 'estoque'

    Returns:
        pd.DataFrame: Produtos premium em estoque
    """
    # === SEU C√ìDIGO AQUI ===

    filtro = (df['preco'] >= 100) & (df['estoque'] > 20)
    return df[filtro]

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de teste
df_produtos = pd.DataFrame({
    'produto': ['Mouse', 'Teclado', 'Monitor', 'Webcam', 'Headset', 'Notebook'],
    'preco': [45.00, 120.00, 890.00, 250.00, 180.00, 3500.00],
    'estoque': [50, 15, 25, 30, 40, 8]
})

print("=== TESTE Q3 ===")
print("\n=== TODOS OS PRODUTOS ===")
print(df_produtos)

# Filtrar produtos premium
premium = filtrar_produtos_premium(df_produtos)

print("\n=== PRODUTOS PREMIUM (pre√ßo >= 100 E estoque > 20) ===")
print(premium)

print(f"\nTotal de produtos premium: {len(premium)} de {len(df_produtos)}")
print(f"Valor m√©dio dos produtos premium: R$ {premium['preco'].mean():,.2f}")

# An√°lise adicional
print("\n=== AN√ÅLISE POR CONDI√á√ÉO ===")
print(f"Produtos com pre√ßo >= 100: {len(df_produtos[df_produtos['preco'] >= 100])}")
print(f"Produtos com estoque > 20: {len(df_produtos[df_produtos['estoque'] > 20])}")
print(f"Produtos que atendem AMBAS: {len(premium)}")

grader.check("q3")
````

# Exerc√≠cio 4: An√°lise Completa

## Descri√ß√£o
Dado um DataFrame com coluna 'valores', retorne um dicion√°rio com:
- 'media': m√©dia dos valores
- 'mediana': mediana dos valores
- 'desvio_padrao': desvio padr√£o
- 'minimo': valor m√≠nimo
- 'maximo': valor m√°ximo
- 'amplitude': m√°ximo - m√≠nimo

## Exemplo
```python
df = pd.DataFrame({'valores': [10, 20, 30, 40, 50]})
analisar_dataset(df)
# {
#   'media': 30.0,
#   'mediana': 30.0,
#   'desvio_padrao': 15.81,
#   'minimo': 10.0,
#   'maximo': 50.0,
#   'amplitude': 40.0
# }
```

## Dicas
- Use `.mean()`, `.median()`, `.std()`, `.min()`, `.max()`
- Amplitude = m√°ximo - m√≠nimo

**Tempo estimado:** 7 minutos

````python
def analisar_dataset(df):
    """
    Calcula estat√≠sticas descritivas da coluna 'valores'.

    Args:
        df (pd.DataFrame): DataFrame com coluna 'valores'

    Returns:
        dict: Dicion√°rio com estat√≠sticas
    """
    # === SEU C√ìDIGO AQUI ===

    serie = df['valores']
    return {
        'media': serie.mean(),
        'mediana': serie.median(),
        'desvio_padrao': serie.std(),
        'minimo': serie.min(),
        'maximo': serie.max(),
        'amplitude': serie.max() - serie.min()
    }

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de notas
np.random.seed(42)
df_notas = pd.DataFrame({
    'aluno': [f'Aluno {i+1}' for i in range(20)],
    'valores': np.random.uniform(5, 10, 20)
})

print("=== TESTE Q4 ===")
print("\n=== DATASET DE NOTAS ===")
print(df_notas.head(10))
print(f"\n... ({len(df_notas)} alunos no total)")

# Analisar dataset
stats = analisar_dataset(df_notas)

print("\n=== ESTAT√çSTICAS DESCRITIVAS ===")
for chave, valor in stats.items():
    print(f"{chave.replace('_', ' ').title()}: {valor:.2f}")

# Visualiza√ß√£o adicional
print("\n=== DISTRIBUI√á√ÉO ===")
media = stats['media']
acima = len(df_notas[df_notas['valores'] > media])
abaixo = len(df_notas[df_notas['valores'] <= media])
print(f"Acima da m√©dia: {acima} alunos ({acima/len(df_notas)*100:.1f}%)")
print(f"Abaixo da m√©dia: {abaixo} alunos ({abaixo/len(df_notas)*100:.1f}%)")

# Verificar usando describe() do Pandas
print("\n=== COMPARA√á√ÉO COM describe() ===")
print(df_notas['valores'].describe())

grader.check("q4")
````
