# === EXEMPLO: BUSCA DE PALAVRA EM LISTA ===

## Descri√ß√£o do Problema
Voc√™ precisa criar uma fun√ß√£o chamada `buscar_palavra` que:
- Recebe dois par√¢metros:
  - `palavra` (string): a palavra que voc√™ quer buscar
  - `frases` (lista de strings): lista de frases onde buscar
- Retorna um n√∫mero inteiro: quantas frases cont√™m a palavra

## Exemplo de Uso
```python
frases = ["python √© legal", "eu amo programar", "python √© poderoso"]
resultado = buscar_palavra("python", frases)
# resultado deve ser 2 (palavra aparece em 2 frases)
```

## Dicas de Implementa√ß√£o
```python
# Como percorrer uma lista:
for frase in lista_de_frases:
    # fazer algo com cada frase
    
# Como verificar se uma palavra est√° em uma string:
if palavra in frase:
    # palavra foi encontrada!
    
# Como contar:

contador += 1  # incrementa o contador
```

````python
# === IMPLEMENTA√á√ÉO DA FUN√á√ÉO ===

def buscar_palavra(palavra, frases):
    """
    Busca quantas frases cont√™m uma determinada palavra.

    Argumentos:
        palavra (str): A palavra a ser buscada
        frases (list): Lista de strings (frases) onde buscar

    Retorna:
        int: N√∫mero de frases que cont√™m a palavra

    Exemplo:
        >>> buscar_palavra("python", ["python √© legal", "java √© legal"])
        1
    """
    # Inicializar contador
    # Come√ßamos com zero porque ainda n√£o encontramos nada
    contador = 0

    # Percorrer cada frase da lista
    # O loop 'for' passa por cada elemento da lista, um por vez
    for frase in frases:
        # Verificar se a palavra est√° presente na frase
        # O operador 'in' retorna True se a palavra est√° na frase
        if palavra in frase:
            # Se encontrou, incrementar o contador
            contador += 1  # Isso √© o mesmo que: contador = contador + 1

    # Retornar o resultado final
    return contador

# IMPORTANTE: Observe que:
# 1. A fun√ß√£o tem um nome descritivo (buscar_palavra)
# 2. Tem documenta√ß√£o explicando o que faz (docstring)
# 3. Os par√¢metros est√£o claramente definidos
# 4. Retorna exatamente o que foi pedido (um int)
````

````python
# === EXEMPLO 1: RETORNAR TIPO ERRADO ===

def buscar_palavra_erro1(palavra, frases):
    """
    VERS√ÉO INCORRETA: Retorna string ao inv√©s de inteiro
    """
    contador = 0
    for frase in frases:
        if palavra in frase:
            contador += 1

    # ERRO: Convertendo para string!
    return str(contador)  # Retorna "3" ao inv√©s de 3

# Testar manualmente
resultado = buscar_palavra_erro1("python", ["python √© legal", "eu amo python"])
print(f"Resultado: {resultado}")
print(f"Tipo: {type(resultado)}")
print(f"√â inteiro? {isinstance(resultado, int)}")  # False!

print("\n" + "="*60)
print("Agora vamos executar o teste autom√°tico...")
print("="*60)
````
