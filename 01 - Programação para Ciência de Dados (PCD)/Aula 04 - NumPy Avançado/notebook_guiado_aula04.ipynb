# === PR√ÅTICA 1: RESHAPE ===

## Objetivo
Usar `reshape()` para reorganizar dimens√µes de array.

## Tarefa
Transformar array 1D em matriz 2D com dimens√µes especificadas.

## Retorno Esperado
Matriz com shape (linhas, colunas).

## Dicas
```python
# Reshape b√°sico
arr = np.arange(12)
matriz = arr.reshape(3, 4)  # 3 linhas, 4 colunas

# Dimens√£o autom√°tica com -1
matriz = arr.reshape(3, -1)  # 3 linhas, ? colunas
```

````python
def reshape_para_matriz(arr, linhas, colunas):
    """
    Transforma array 1D em matriz 2D.

    Args:
        arr (np.ndarray): Array 1D
        linhas (int): N√∫mero de linhas
        colunas (int): N√∫mero de colunas

    Returns:
        np.ndarray: Matriz com shape (linhas, colunas)
    """
    # === SEU C√ìDIGO AQUI ===

    matriz = arr.reshape(linhas, colunas)

    return matriz

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar fun√ß√£o
arr = np.arange(12)
print(f"Array original: {arr}")
print(f"Shape: {arr.shape}")

# Reshape para diferentes dimens√µes
print("\n=== RESHAPE PARA DIFERENTES FORMAS ===")

matriz_3x4 = reshape_para_matriz(arr, 3, 4)
print(f"\n3x4:\n{matriz_3x4}")
print(f"Shape: {matriz_3x4.shape}")

matriz_2x6 = reshape_para_matriz(arr, 2, 6)
print(f"\n2x6:\n{matriz_2x6}")
print(f"Shape: {matriz_2x6.shape}")

matriz_4x3 = reshape_para_matriz(arr, 4, 3)
print(f"\n4x3:\n{matriz_4x3}")
print(f"Shape: {matriz_4x3.shape}")

# Verificar teste
grader.check("p1")
````

# === PR√ÅTICA 2: TRANSPOSE ===

## Objetivo
Transpor matriz (trocar linhas por colunas).

## Tarefa
Implementar transposi√ß√£o de matriz.

## Retorno Esperado
Matriz transposta.

## Dicas
```python
# Transpor matriz
matriz = np.array([[1, 2], [3, 4]])
transposta = matriz.T

# Ou usando fun√ß√£o
transposta = np.transpose(matriz)
```

````python
def transpor_matriz(matriz):
    """
    Transp√µe uma matriz.

    Args:
        matriz (np.ndarray): Matriz 2D

    Returns:
        np.ndarray: Matriz transposta
    """
    # === SEU C√ìDIGO AQUI ===

    transposta = matriz.T

    return transposta

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar matriz de exemplo
matriz = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("=== TRANSPOSI√á√ÉO DE MATRIZ ===")
print(f"\nOriginal ({matriz.shape}):")
print(matriz)

# Transpor
transposta = transpor_matriz(matriz)

print(f"\nTransposta ({transposta.shape}):")
print(transposta)

# Verificar propriedade: (AT)T = A
dupla_transposta = transpor_matriz(transposta)
print(f"\nDupla transposta volta ao original:")
print(np.array_equal(matriz, dupla_transposta))

# Verificar teste
grader.check("p2")
````

# === PR√ÅTICA 3: PRODUTO MATRICIAL ===

## Objetivo
Realizar produto matricial (multiplica√ß√£o de matrizes).

## Tarefa
Implementar multiplica√ß√£o de duas matrizes.

## Retorno Esperado
Resultado do produto matricial A @ B.

## Dicas
```python
# Produto matricial
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Tr√™s formas equivalentes
C = A @ B              # Operador @ (preferido)
C = np.dot(A, B)       # Fun√ß√£o dot
C = np.matmul(A, B)    # Fun√ß√£o matmul
```

````python
def produto_matricial(A, B):
    """
    Calcula produto matricial de A e B.

    Args:
        A (np.ndarray): Primeira matriz
        B (np.ndarray): Segunda matriz

    Returns:
        np.ndarray: Produto matricial A @ B
    """
    # === SEU C√ìDIGO AQUI ===

    resultado = A @ B

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Exemplo de multiplica√ß√£o de matrizes
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("=== PRODUTO MATRICIAL ===")
print(f"\nA ({A.shape}):")
print(A)
print(f"\nB ({B.shape}):")
print(B)

# Calcular produto
C = produto_matricial(A, B)

print(f"\nA @ B ({C.shape}):")
print(C)

# Comparar com multiplica√ß√£o elemento-a-elemento
print("\n=== DIFEREN√áA: @ vs * ===")
print(f"A @ B (matricial):\n{C}")
print(f"\nA * B (elemento-a-elemento):\n{A * B}")

# Verificar teste
grader.check("p3")
````

# === PR√ÅTICA 4: INVERSA DE MATRIZ ===

## Objetivo
Calcular inversa de uma matriz.

## Tarefa
Implementar c√°lculo da inversa usando `np.linalg.inv()`.

## Retorno Esperado
Matriz inversa tal que A @ A_inv = I (identidade).

## Dicas
```python
# Calcular inversa
A = np.array([[1, 2], [3, 4]], dtype=float)
A_inv = np.linalg.inv(A)

# Verificar: A @ A_inv = I
I = A @ A_inv
print(I)  # Deve ser aproximadamente [[1, 0], [0, 1]]
```

````python
def calcular_inversa(A):
    """
    Calcula a inversa de uma matriz.

    Args:
        A (np.ndarray): Matriz quadrada

    Returns:
        np.ndarray: Matriz inversa
    """
    # === SEU C√ìDIGO AQUI ===

    A_inv = np.linalg.inv(A)

    return A_inv

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar matriz
A = np.array([[1, 2],
              [3, 4]], dtype=float)

print("=== INVERSA DE MATRIZ ===")
print(f"\nMatriz A:")
print(A)

# Calcular inversa
A_inv = calcular_inversa(A)

print(f"\nMatriz A^-1:")
print(A_inv)

# Verificar: A @ A_inv = I
I = A @ A_inv
print(f"\nA @ A^-1 (deve ser identidade):")
print(I)

# Verificar se √© aproximadamente identidade
I_esperada = np.eye(2)
print(f"\n√â identidade? {np.allclose(I, I_esperada)}")

# Verificar teste
grader.check("p4")
````

# === PR√ÅTICA 5: RESOLVER SISTEMA LINEAR ===

## Objetivo
Resolver sistema de equa√ß√µes lineares Ax = b.

## Tarefa
Usar `np.linalg.solve()` para encontrar x.

## Retorno Esperado
Vetor x que satisfaz Ax = b.

## Dicas
```python
# Sistema: 2x + y = 8
#          x - y = 1
A = np.array([[2, 1], [1, -1]], dtype=float)
b = np.array([8, 1], dtype=float)

# Resolver
x = np.linalg.solve(A, b)
# x = [3, 2] (x=3, y=2)
```

````python
def resolver_sistema(A, b):
    """
    Resolve sistema linear Ax = b.

    Args:
        A (np.ndarray): Matriz de coeficientes
        b (np.ndarray): Vetor de termos independentes

    Returns:
        np.ndarray: Vetor solu√ß√£o x
    """
    # === SEU C√ìDIGO AQUI ===

    x = np.linalg.solve(A, b)

    return x

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Sistema de equa√ß√µes
# 2x + y = 8
# x - y = 1
A = np.array([[2, 1],
              [1, -1]], dtype=float)
b = np.array([8, 1], dtype=float)

print("=== RESOLVER SISTEMA LINEAR ===")
print("\nSistema:")
print("2x + y = 8")
print("x - y = 1")

# Resolver
x = resolver_sistema(A, b)

print(f"\nSolu√ß√£o:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")

# Verificar solu√ß√£o
resultado = A @ x
print(f"\nVerifica√ß√£o:")
print(f"Ax = {resultado}")
print(f"b  = {b}")
print(f"Ax = b? {np.allclose(resultado, b)}")

# Verificar teste
grader.check("p5")
````

# === PR√ÅTICA 6: CONCATENA√á√ÉO VERTICAL ===

## Objetivo
Concatenar arrays verticalmente (empilhar linhas).

## Tarefa
Usar `np.vstack()` ou `np.concatenate()` com axis=0.

## Retorno Esperado
Array resultante da concatena√ß√£o vertical.

## Dicas
```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Duas formas equivalentes
c = np.vstack([a, b])
c = np.concatenate([a, b], axis=0)
```

````python
def concatenar_vertical(a, b):
    """
    Concatena dois arrays verticalmente.

    Args:
        a (np.ndarray): Primeiro array
        b (np.ndarray): Segundo array

    Returns:
        np.ndarray: Array concatenado verticalmente
    """
    # === SEU C√ìDIGO AQUI ===

    resultado = np.vstack([a, b])

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Arrays de exemplo
a = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([[7, 8, 9],
              [10, 11, 12]])

print("=== CONCATENA√á√ÉO VERTICAL ===")
print(f"\nArray A ({a.shape}):")
print(a)
print(f"\nArray B ({b.shape}):")
print(b)

# Concatenar
resultado = concatenar_vertical(a, b)

print(f"\nResultado ({resultado.shape}):")
print(resultado)

# Aplica√ß√£o pr√°tica: adicionar novas amostras
print("\nüí° Uso pr√°tico: adicionar novos dados ao dataset")
print("   Dataset original: 2 amostras")
print("   Novos dados: 2 amostras")
print("   Dataset final: 4 amostras")

# Verificar teste
grader.check("p6")
````

# === PR√ÅTICA 7: STACK (EMPILHAR) ===

## Objetivo
Empilhar arrays criando nova dimens√£o.

## Tarefa
Usar `np.stack()` para criar novo eixo.

## Retorno Esperado
Array com dimens√£o adicional.

## Dicas
```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Stack ao longo de novo eixo 0
c = np.stack([a, b], axis=0)
# [[1, 2, 3],
#  [4, 5, 6]]
```

````python
def empilhar_arrays(a, b):
    """
    Empilha dois arrays criando nova dimens√£o.

    Args:
        a (np.ndarray): Primeiro array
        b (np.ndarray): Segundo array

    Returns:
        np.ndarray: Arrays empilhados
    """
    # === SEU C√ìDIGO AQUI ===

    resultado = np.stack([a, b], axis=0)

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Vetores 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("=== STACK: CRIAR NOVA DIMENS√ÉO ===")
print(f"\nArray A ({a.shape}): {a}")
print(f"Array B ({b.shape}): {b}")

# Empilhar
resultado = empilhar_arrays(a, b)

print(f"\nResultado ({resultado.shape}):")
print(resultado)

# Comparar com vstack
print("\n=== DIFEREN√áA: stack vs vstack ===")
print(f"stack (cria nova dim): shape {resultado.shape}")
vstack_resultado = np.vstack([a, b])
print(f"vstack (usa dim exist): shape {vstack_resultado.shape}")
print(f"\nResultados iguais? {np.array_equal(resultado, vstack_resultado)}")

# Verificar teste
grader.check("p7")
````

# === PR√ÅTICA 8: SALVAR E CARREGAR ===

## Objetivo
Salvar array em disco e carregar novamente.

## Tarefa
Usar `np.save()` para salvar e `np.load()` para carregar.

## Retorno Esperado
Array carregado id√™ntico ao original.

## Dicas
```python
# Salvar
arr = np.array([1, 2, 3, 4, 5])
np.save('dados.npy', arr)

# Carregar
arr_loaded = np.load('dados.npy')

# Verificar igualdade
np.array_equal(arr, arr_loaded)  # True
```

````python
def salvar_carregar_array(arr, filename):
    """
    Salva array em disco e carrega de volta.

    Args:
        arr (np.ndarray): Array para salvar
        filename (str): Nome do arquivo

    Returns:
        np.ndarray: Array carregado
    """
    # === SEU C√ìDIGO AQUI ===

    # Salvar
    np.save(filename, arr)

    # Carregar
    arr_loaded = np.load(filename)

    return arr_loaded

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar array de exemplo
arr_original = np.random.randn(5, 3)

print("=== SALVAR E CARREGAR ARRAYS ===")
print(f"\nArray original:")
print(arr_original)
print(f"Shape: {arr_original.shape}")
print(f"Dtype: {arr_original.dtype}")

# Salvar e carregar
arr_loaded = salvar_carregar_array(arr_original, 'test_array.npy')

print(f"\nArray carregado:")
print(arr_loaded)
print(f"Shape: {arr_loaded.shape}")
print(f"Dtype: {arr_loaded.dtype}")

# Verificar igualdade
print(f"\n=== VERIFICA√á√ÉO ===")
print(f"Arrays s√£o iguais? {np.array_equal(arr_original, arr_loaded)}")
print(f"Shapes iguais? {arr_original.shape == arr_loaded.shape}")
print(f"Dtypes iguais? {arr_original.dtype == arr_loaded.dtype}")

# Verificar se arquivo foi criado
import os
print(f"\nArquivo existe? {os.path.exists('test_array.npy')}")
if os.path.exists('test_array.npy'):
    size = os.path.getsize('test_array.npy')
    print(f"Tamanho do arquivo: {size} bytes")

# Verificar teste
grader.check("p8")
````
