# Exerc√≠cio 1: Reshape para 3D

## Descri√ß√£o
Transforme um array 1D em um tensor 3D com as dimens√µes especificadas.
Isso √© comum ao trabalhar com batches de imagens ou s√©ries temporais.

## Exemplo
```python
arr = np.arange(24)  # 24 elementos
reshape_3d(arr, 2, 3, 4)  # (2, 3, 4)
# 2 batches, 3 linhas, 4 colunas
```

## Requisitos
- Use `reshape()`
- Verifique se o n√∫mero total de elementos √© compat√≠vel

**Tempo estimado:** 5 minutos

````python
def reshape_3d(arr, dim1, dim2, dim3):
    """
    Reshape array 1D para 3D.

    Args:
        arr (np.ndarray): Array 1D
        dim1, dim2, dim3 (int): Dimens√µes desejadas

    Returns:
        np.ndarray: Array 3D com shape (dim1, dim2, dim3)
    """
    # === SEU C√ìDIGO AQUI ===

    if arr.size != dim1 * dim2 * dim3:
        raise ValueError("N√∫mero total de elementos incompat√≠vel com as dimens√µes fornecidas.")
    return arr.reshape(dim1, dim2, dim3)

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q1 ===")
arr = np.arange(24)
print(f"Array 1D: shape {arr.shape}")
print(arr)

tensor = reshape_3d(arr, 2, 3, 4)
print(f"\nTensor 3D: shape {tensor.shape}")
print(tensor)
print(f"\nPrimeiro batch (2D):")
print(tensor[0])

grader.check("q1")
````

# Exerc√≠cio 2: Autovalores e Autovetores

## Descri√ß√£o
Calcule os autovalores e autovetores de uma matriz quadrada.
Isso √© fundamental para PCA, an√°lise de estabilidade, e outras aplica√ß√µes.

## Conceito
Para uma matriz A e autovetor v:
- A @ v = Œª * v
- Œª √© o autovalor correspondente

## Exemplo
```python
A = np.array([[2, 0], [0, 3]])
eigenvalues, eigenvectors = calcular_autovalores(A)
# eigenvalues: [2, 3]
# eigenvectors: [[1, 0], [0, 1]]
```

## Dica
Use `np.linalg.eig()`

**Tempo estimado:** 7 minutos

````python
def calcular_autovalores(A):
    """
    Calcula autovalores e autovetores.

    Args:
        A (np.ndarray): Matriz quadrada

    Returns:
        tuple: (eigenvalues, eigenvectors)
    """
    # === SEU C√ìDIGO AQUI ===

    eigenvalues, eigenvectors = np.linalg.eig(A)
    return eigenvalues, eigenvectors

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q2 ===")
A = np.array([[4, 2],
              [2, 3]], dtype=float)
print("Matriz A:")
print(A)

eigenvalues, eigenvectors = calcular_autovalores(A)
print(f"\nAutovalores: {eigenvalues}")
print(f"\nAutovetores:")
print(eigenvectors)

# Verificar propriedade: A @ v = Œª * v
print("\n=== VERIFICA√á√ÉO ===")
v = eigenvectors[:, 0]
lam = eigenvalues[0]
print(f"A @ v = {A @ v}")
print(f"Œª * v = {lam * v}")
print(f"S√£o iguais? {np.allclose(A @ v, lam * v)}")

grader.check("q2")
````

# Exerc√≠cio 3: Concatenar Features e Labels

## Descri√ß√£o
Em Machine Learning, √†s vezes precisamos combinar a matriz de features (X)
com o vetor de labels (y) em uma √∫nica matriz.

Adicione a coluna de labels como √∫ltima coluna da matriz de features.

## Exemplo
```python
X = np.array([[1, 2],    # 2 amostras, 2 features
              [3, 4]])
y = np.array([[0],       # 2 labels
              [1]])
              
adicionar_coluna_labels(X, y)
# [[1, 2, 0],
#  [3, 4, 1]]
```

## Dicas
- Use `np.hstack()` ou `np.column_stack()` ou `np.c_[]`
- Certifique-se que y tem shape (n, 1) e n√£o (n,)

**Tempo estimado:** 8 minutos

````python
def adicionar_coluna_labels(X, y):
    """
    Adiciona coluna de labels √†s features.

    Args:
        X (np.ndarray): Matriz de features (n, m)
        y (np.ndarray): Vetor de labels (n, 1) ou (n,)

    Returns:
        np.ndarray: Matriz combinada (n, m+1)
    """
    # === SEU C√ìDIGO AQUI ===

    if y.ndim == 1:
        y = y.reshape(-1, 1)
    return np.hstack((X, y))

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q3 ===")

# Exemplo 1: y como (n, 1)
X = np.array([[1.5, 2.3],
              [3.2, 4.1],
              [5.0, 6.8]])
y = np.array([[0],
              [1],
              [1]])

print("Features X:")
print(X)
print(f"Shape: {X.shape}")

print("\nLabels y:")
print(y)
print(f"Shape: {y.shape}")

resultado = adicionar_coluna_labels(X, y)
print("\nCombinado:")
print(resultado)
print(f"Shape: {resultado.shape}")

# Exemplo 2: y como (n,)
print("\n=== TESTE COM y 1D ===")
y_1d = np.array([0, 1, 1])
resultado2 = adicionar_coluna_labels(X, y_1d)
print(f"Shape resultado: {resultado2.shape}")
print(resultado2)

grader.check("q3")
````

# Exerc√≠cio 4: Dist√¢ncia Euclidiana

## Descri√ß√£o
Calcule a dist√¢ncia euclidiana entre dois pontos.
Essa m√©trica √© fundamental para k-NN, clustering, e outros algoritmos.

## F√≥rmula
Para pontos p e q:
```
dist = ‚àö(Œ£(pi - qi)¬≤)
```

## Exemplo
```python
p1 = np.array([0, 0])
p2 = np.array([3, 4])
calcular_distancia_euclidiana(p1, p2)
# 5.0  (tri√¢ngulo 3-4-5)
```

## Dicas
- Use `np.linalg.norm()` ou
- Calcule manualmente: `np.sqrt(np.sum((p1 - p2)**2))`

**Tempo estimado:** 5 minutos

````python
def calcular_distancia_euclidiana(p1, p2):
    """
    Calcula dist√¢ncia euclidiana entre dois pontos.

    Args:
        p1 (np.ndarray): Primeiro ponto
        p2 (np.ndarray): Segundo ponto

    Returns:
        float: Dist√¢ncia euclidiana
    """
    # === SEU C√ìDIGO AQUI ===

    return np.linalg.norm(p1 - p2)

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
print("=== TESTE Q4 ===")

# Teste 1: Tri√¢ngulo 3-4-5
p1 = np.array([0, 0])
p2 = np.array([3, 4])
dist = calcular_distancia_euclidiana(p1, p2)
print(f"p1 = {p1}")
print(f"p2 = {p2}")
print(f"Dist√¢ncia: {dist} (esperado: 5.0)")

# Teste 2: Pontos 3D
print("\n=== TESTE 3D ===")
p3 = np.array([1, 2, 3])
p4 = np.array([4, 6, 8])
dist2 = calcular_distancia_euclidiana(p3, p4)
print(f"p3 = {p3}")
print(f"p4 = {p4}")
print(f"Dist√¢ncia: {dist2:.2f}")

# Teste 3: Mesmo ponto
print("\n=== TESTE DIST√ÇNCIA ZERO ===")
p5 = np.array([5, 5])
dist3 = calcular_distancia_euclidiana(p5, p5)
print(f"Dist√¢ncia para si mesmo: {dist3} (esperado: 0.0)")

grader.check("q4")
````
