# Quest√£o 3: Regress√£o Linear - Equa√ß√£o Normal (25 pontos)

## Contexto
A regress√£o linear √© fundamental em Machine Learning.
Voc√™ vai implement√°-la do zero usando √°lgebra linear.

## Teoria: Equa√ß√£o Normal
Para encontrar os par√¢metros Œ∏ que minimizam o erro:

```
Œ∏ = (X^T X)^-1 X^T y
```

Onde:
- X √© a matriz de features (com coluna de 1s para intercepto)
- y √© o vetor de targets
- Œ∏ cont√©m [intercepto, coef1, coef2, ...]

## Tarefa
Implemente regress√£o linear que:
1. Adiciona coluna de 1s ao in√≠cio de X (para intercepto)
2. Calcula Œ∏ usando a equa√ß√£o normal
3. Separa intercepto e coeficientes
4. Calcula R¬≤ (coeficiente de determina√ß√£o)
5. Retorna dicion√°rio com todos os resultados

## F√≥rmula do R¬≤
```
R¬≤ = 1 - (SS_res / SS_tot)

onde:
SS_res = Œ£(y - y_pred)¬≤  # Soma dos quadrados dos res√≠duos
SS_tot = Œ£(y - y_mean)¬≤  # Soma total dos quadrados
```

## Exemplo
```python
# Dados: y = 2*x1 + 3*x2 + 5 + ru√≠do

y = np.array([12, 17, 22])  # aproximadamente
```
