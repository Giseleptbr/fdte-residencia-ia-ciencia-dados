# === PR√ÅTICA 1: MATRIZ DE CORRELA√á√ÉO ===

## Objetivo
Calcular matriz de correla√ß√£o entre vari√°veis num√©ricas.

## Tarefa
Dado DataFrame, calcule a matriz de correla√ß√£o usando o
m√©todo de Pearson para todas as colunas num√©ricas.

## Retorno Esperado
DataFrame com matriz de correla√ß√£o.

## Dicas
```python
# Selecionar apenas colunas num√©ricas
df_numeric = df.select_dtypes(include=[np.number])

# Calcular correla√ß√£o
corr_matrix = df_numeric.corr(method='pearson')
```

## Interpreta√ß√£o
- Correla√ß√£o pr√≥xima de +1: forte rela√ß√£o positiva
- Correla√ß√£o pr√≥xima de -1: forte rela√ß√£o negativa
- Correla√ß√£o pr√≥xima de 0: sem rela√ß√£o linear

**Tempo estimado:** 5 minutos

````python
def calcular_matriz_correlacao(df):
    """
    Calcula matriz de correla√ß√£o de Pearson.
    
    Args:
        df (pd.DataFrame): DataFrame com vari√°veis num√©ricas
        
    Returns:
        pd.DataFrame: Matriz de correla√ß√£o
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Selecionar apenas colunas num√©ricas
    df_numeric = df.select_dtypes(include=[np.number])
    
    # Calcular correla√ß√£o
    corr_matrix = df_numeric.corr(method='pearson')
    
    return corr_matrix
    
    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar DataFrame de exemplo
np.random.seed(42)
n = 100

df_exemplo = pd.DataFrame({
    'idade': np.random.randint(20, 60, n),
    'salario': np.random.normal(5000, 2000, n),
    'anos_experiencia': np.random.randint(0, 30, n),
    'categoria': np.random.choice(['A', 'B', 'C'], n)  # Categ√≥rica
})

# Adicionar correla√ß√£o artificial entre sal√°rio e experi√™ncia
df_exemplo['salario'] = df_exemplo['salario'] + df_exemplo['anos_experiencia'] * 100

print("\n=== DADOS DE EXEMPLO ===")
print(df_exemplo.head())

# Calcular correla√ß√£o
corr_matrix = calcular_matriz_correlacao(df_exemplo)

print("\n=== MATRIZ DE CORRELA√á√ÉO ===")
print(corr_matrix.round(3))

print("\n=== INTERPRETA√á√ÉO ===")
print("Valores pr√≥ximos a:")
print("  +1.0: correla√ß√£o positiva forte")
print("   0.0: sem correla√ß√£o linear")
print("  -1.0: correla√ß√£o negativa forte")

# Verificar teste
grader.check("p1")
````

````python
def criar_heatmap_correlacao(df):
    """
    Cria heatmap de correla√ß√£o.
    
    Args:
        df (pd.DataFrame): DataFrame com vari√°veis num√©ricas
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Selecionar apenas num√©ricas e calcular correla√ß√£o
    df_numeric = df.select_dtypes(include=[np.number])
    corr = df_numeric.corr()
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Criar heatmap
    sns.heatmap(corr, 
                annot=True, 
                fmt='.2f',
                cmap='coolwarm', 
                vmin=-1, 
                vmax=1, 
                center=0,
                square=True,
                linewidths=1,
                cbar_kws={'label': 'Correla√ß√£o'},
                ax=ax)
    
    ax.set_title('Heatmap de Correla√ß√£o', fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 3: SCATTER PLOT ===

## Objetivo

## Tarefa
Dados DataFrame e nomes de duas colunas (col_x, col_y),
crie scatter plot com:
- Pontos com transpar√™ncia (alpha=0.6)
- Linha de tend√™ncia (regress√£o linear)
- Valor de correla√ß√£o no t√≠tulo

## Retorno Esperado

## Dicas
```python
# Scatter plot
ax.scatter(df[col_x], df[col_y], alpha=0.6)

z = np.polyfit(df[col_x], df[col_y], 1)
```

**Tempo estimado:** 7 minutos

````python
def criar_scatter_plot(df, col_x, col_y):
    """
    Cria scatter plot com linha de tend√™ncia.
    
    Args:
        df (pd.DataFrame): DataFrame
        col_x (str): Nome da coluna X
        col_y (str): Nome da coluna Y
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Scatter plot
    ax.scatter(df[col_x], df[col_y], alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
    
    # Linha de tend√™ncia
    z = np.polyfit(df[col_x], df[col_y], 1)
    p = np.poly1d(z)
    ax.plot(df[col_x].sort_values(), p(df[col_x].sort_values()), 
            'r--', linewidth=2, label='Linha de tend√™ncia')
    
    # Calcular correla√ß√£o
    corr = df[col_x].corr(df[col_y])
    
    # Labels e t√≠tulo
    ax.set_xlabel(col_x, fontsize=12)
    ax.set_ylabel(col_y, fontsize=12)
    ax.set_title(f'Scatter Plot: {col_x} vs {col_y}\nCorrela√ß√£o: {corr:.3f}', 
                 fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU C√ìDIGO ===
````

````python
# Criar scatter plot
print("\n=== SCATTER PLOT ===")

fig, ax = criar_scatter_plot(df_exemplo, 'anos_experiencia', 'salario')
plt.show()

print("\n=== INTERPRETA√á√ÉO ===")
print("A linha vermelha tracejada mostra a tend√™ncia linear.")
print("Quanto mais pr√≥ximos os pontos estiverem da linha, maior a correla√ß√£o.")

# Verificar teste
grader.check("p3")
````

# === PR√ÅTICA 4: ENCONTRAR CORRELA√á√ïES FORTES ===

## Objetivo
Identificar pares de vari√°veis com correla√ß√£o forte.

## Tarefa
Dado DataFrame e threshold (padr√£o 0.7),
retorne lista de tuplas (var1, var2, correla√ß√£o)
onde |correla√ß√£o| >= threshold.

Exclua correla√ß√µes da vari√°vel com ela mesma (diag).

## Retorno Esperado
Lista de tuplas [(var1, var2, corr), ...].

## Dicas
```python
# Matriz de correla√ß√£o
corr = df.corr()

# Pegar apenas tri√¢ngulo superior (evitar duplicatas)
upper_tri = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# Encontrar correla√ß√µes fortes
strong_corr = [(col, idx, upper_tri.loc[idx, col]) 
               for col in upper_tri.columns 
               for idx in upper_tri.index 
               if abs(upper_tri.loc[idx, col]) >= threshold]
```

**Tempo estimado:** 7 minutos

````python
def encontrar_correlacoes_fortes(df, threshold=0.7):
    """
    Encontra pares de vari√°veis com correla√ß√£o forte.
    
    Args:
        df (pd.DataFrame): DataFrame
        threshold (float): Limite m√≠nimo (padr√£o 0.7)
        
    Returns:
        list: Lista de tuplas (var1, var2, correla√ß√£o)
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Selecionar num√©ricas e calcular correla√ß√£o
    df_numeric = df.select_dtypes(include=[np.number])
    corr = df_numeric.corr()
    
    # Pegar apenas tri√¢ngulo superior (k=1 exclui diagonal)
    upper_tri = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    
    # Encontrar correla√ß√µes fortes
    strong_corr = []
    for col in upper_tri.columns:
        for idx in upper_tri.index:
            valor = upper_tri.loc[idx, col]
            if pd.notna(valor) and abs(valor) >= threshold:
                strong_corr.append((idx, col, valor))
    
    # Ordenar por valor absoluto (maior primeiro)
    strong_corr.sort(key=lambda x: abs(x[2]), reverse=True)
    
    return strong_corr
    
    # === FIM DO SEU C√ìDIGO ===
````

````python
# Encontrar correla√ß√µes fortes
print("\n=== BUSCANDO CORRELA√á√ïES FORTES (>= 0.7) ===")

correlacoes_fortes = encontrar_correlacoes_fortes(df_exemplo, threshold=0.7)

if correlacoes_fortes:
    print(f"\nEncontradas {len(correlacoes_fortes)} correla√ß√µes fortes:\n")
    for var1, var2, corr in correlacoes_fortes:
        print(f"  {var1} ‚Üî {var2}: {corr:+.3f}")
else:
    print("\nNenhuma correla√ß√£o forte encontrada com threshold >= 0.7")

# Tentar com threshold menor
print("\n=== BUSCANDO CORRELA√á√ïES MODERADAS (>= 0.5) ===")
correlacoes_moderadas = encontrar_correlacoes_fortes(df_exemplo, threshold=0.5)

if correlacoes_moderadas:
    print(f"\nEncontradas {len(correlacoes_moderadas)} correla√ß√µes moderadas/fortes:\n")
    for var1, var2, corr in correlacoes_moderadas:
        print(f"  {var1} ‚Üî {var2}: {corr:+.3f}")

print("\n=== CLASSIFICA√á√ÉO ===")
print("  0.0 - 0.3: correla√ß√£o fraca")
print("  0.3 - 0.7: correla√ß√£o moderada")
print("  0.7 - 1.0: correla√ß√£o forte")

# Verificar teste
grader.check("p4")
````

# === PR√ÅTICA 5: PAIRPLOT ===

## Objetivo
Criar pairplot para visualizar rela√ß√µes entre m√∫ltiplas vari√°veis.

## Tarefa
Dado DataFrame, crie pairplot usando Seaborn.

Pairplot mostra:
- Scatter plots para cada par de vari√°veis
- Histogramas na diagonal

## Retorno Esperado
Objeto PairGrid do Seaborn.

## Dicas
```python
# Selecionar apenas colunas num√©ricas
df_numeric = df.select_dtypes(include=[np.number])

# Criar pairplot

```

**Tempo estimado:** 5 minutos

````python
def criar_pairplot(df):
    """
    Cria pairplot de vari√°veis num√©ricas.
    
    Args:
        df (pd.DataFrame): DataFrame
        
    Returns:
        seaborn.PairGrid: Pairplot
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Selecionar apenas num√©ricas
    df_numeric = df.select_dtypes(include=[np.number])
    
    # Criar pairplot
    pairplot = sns.pairplot(
        df_numeric, 
        diag_kind='hist',
        plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'black', 'linewidth': 0.5},
        diag_kws={'bins': 20, 'edgecolor': 'black'}
    )
    
    pairplot.fig.suptitle('Pairplot - An√°lise Multivariada', 
                          fontsize=16, fontweight='bold', y=1.01)
    
    return pairplot
    
    # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 6: AN√ÅLISE DE VARI√ÅVEL CATEG√ìRICA ===

## Objetivo
Visualizar distribui√ß√£o de vari√°vel categ√≥rica.

## Tarefa
Dado DataFrame e nome de coluna categ√≥rica,

- Percentuais

## Retorno Esperado

```python
# Contar valores
counts = df[col].value_counts()

# Criar bar plot
```

**Tempo estimado:** 6 minutos

````python
def analisar_variaveis_categoricas(df, col):
    """
    Cria gr√°fico de barras para vari√°vel categ√≥rica.
    
    Args:
        df (pd.DataFrame): DataFrame
        col (str): Nome da coluna categ√≥rica
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU C√ìDIGO AQUI ===
    
    # Contar valores
    counts = df[col].value_counts()
    percentages = df[col].value_counts(normalize=True) * 100
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Criar bar plot
    bars = ax.bar(range(len(counts)), counts.values, 
                   color='steelblue', edgecolor='black', alpha=0.7)
    
    # Configurar eixo X
    ax.set_xticks(range(len(counts)))
    ax.set_xticklabels(counts.index)
    
    # Labels
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Frequ√™ncia', fontsize=12)
    ax.set_title(f'Distribui√ß√£o de {col}', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y')
    
    # Adicionar valores e percentuais
    for i, (cat, count) in enumerate(counts.items()):
        pct = percentages[cat]
        ax.text(i, count + max(counts) * 0.02, 
                f'{count}\n({pct:.1f}%)', 
                ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU C√ìDIGO ===
````

````python
# Analisar vari√°vel categ√≥rica
print("\n=== AN√ÅLISE DE VARI√ÅVEL CATEG√ìRICA ===")

fig, ax = analisar_variaveis_categoricas(df_exemplo, 'categoria')
plt.show()

print("\n=== ESTAT√çSTICAS ===")
print(df_exemplo['categoria'].value_counts())
print("\nPercentuais:")
print(df_exemplo['categoria'].value_counts(normalize=True) * 100)

# Verificar teste
grader.check("p6")
````
