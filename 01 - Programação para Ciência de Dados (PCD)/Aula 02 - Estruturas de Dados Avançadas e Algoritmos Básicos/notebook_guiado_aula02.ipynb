# === PR√ÅTICA 1: LISTAS ANINHADAS (Bloco 1) ===

## Objetivo
Criar uma estrutura de lista aninhada que agrupe passageiros por classe.

## Tarefa
Criar uma **lista de 3 sublistas**, onde:
- `resultado[0]` = lista com todos os passageiros da classe 1
- `resultado[1]` = lista com todos os passageiros da classe 2
- `resultado[2]` = lista com todos os passageiros da classe 3

## Estrutura Esperada
```python
[
  [passageiro1_classe1, passageiro2_classe1, ...],  # Classe 1
  [passageiro1_classe2, passageiro2_classe2, ...],  # Classe 2
  [passageiro1_classe3, passageiro2_classe3, ...]   # Classe 3
]
```

## Dicas
```python
# Criar 3 listas vazias
classe1, classe2, classe3 = [], [], []

# Ou criar lista de listas
resultado = [[], [], []]

# Adicionar passageiro na lista correta
if passageiro['Pclass'] == 1:
    resultado[0].append(passageiro)
```

````python
def agrupar_por_classe(dados):
    """
    Agrupa passageiros em listas por classe.

    Args:
        dados (list): Lista de dicion√°rios com dados dos passageiros

    Returns:
        list: Lista com 3 sublistas [classe1, classe2, classe3]
    """
    # === SEU C√ìDIGO AQUI ===

    # Criar lista com 3 sublistas vazias
    resultado = [[], [], []]

    # Iterar sobre passageiros
    for passageiro in dados:
        # Pegar n√∫mero da classe
        classe = passageiro['Pclass']

        # Adicionar √† sublista correspondente (classe - 1 porque √≠ndice come√ßa em 0)
        resultado[classe - 1].append(passageiro)

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
agrupados = agrupar_por_classe(passageiros)

print("\n=== PASSAGEIROS AGRUPADOS POR CLASSE ===")
print(f"\nClasse 1: {len(agrupados[0])} passageiros")
for p in agrupados[0]:
    print(f"  - {p['Name']}")

print(f"\nClasse 2: {len(agrupados[1])} passageiros")
for p in agrupados[1]:
    print(f"  - {p['Name']}")

print(f"\nClasse 3: {len(agrupados[2])} passageiros")
for p in agrupados[2]:
    print(f"  - {p['Name']}")

# Verificar teste
grader.check("p1")
````

# === PR√ÅTICA 2: COUNTER (Bloco 1) ===

## Objetivo
Usar `collections.Counter` para contar passageiros por sexo.

## Tarefa
Contar quantos passageiros s√£o 'male' e quantos s√£o 'female'.

## Retorno Esperado
```python
Counter({'male': X, 'female': Y})
# ou dicion√°rio normal
{'male': X, 'female': Y}
```

## Dicas
```python
# Importar Counter
from collections import Counter

# Extrair lista de sexos
sexos = [p['Sex'] for p in dados]

# Contar com Counter
contagem = Counter(sexos)

# Ou fazer manualmente
contagem = {}
for p in dados:
    sexo = p['Sex']
    if sexo not in contagem:
        contagem[sexo] = 0
    contagem[sexo] += 1
```

````python
from collections import Counter

def contar_por_sexo(dados):
    """
    Conta passageiros por sexo usando Counter.

    Args:
        dados (list): Lista de dicion√°rios com dados dos passageiros

    Returns:
        Counter ou dict: Contagem de passageiros por sexo
    """
    # === SEU C√ìDIGO AQUI ===

    # Extrair lista de sexos usando list comprehension
    sexos = [p['Sex'] for p in dados]

    # Usar Counter para contar
    contagem = Counter(sexos)

    return contagem

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar sua fun√ß√£o
contagem = contar_por_sexo(passageiros)

print("\n=== CONTAGEM POR SEXO ===")
print(f"Homens: {contagem['male']}")
print(f"Mulheres: {contagem['female']}")
print(f"\nTotal: {sum(contagem.values())} passageiros")

# Verificar teste
grader.check("p2")
````

# === PR√ÅTICA 3: LIST COMPREHENSION (Bloco 1) ===

## Objetivo
Usar list comprehension para filtrar dados.

## Tarefa
Filtrar apenas os passageiros que **sobreviveram** (Survived == 1).

## Retorno Esperado
Lista com apenas os dicion√°rios dos sobreviventes.

## Dicas
```python
# List comprehension com filtro
resultado = [item for item in lista if condicao]

# Exemplo
pares = [x for x in numeros if x % 2 == 0]

# Para nosso caso
sobreviventes = [p for p in dados if p['Survived'] == 1]
```

````python
def filtrar_sobreviventes(dados):
    """
    Filtra apenas passageiros que sobreviveram.

    Args:
        dados (list): Lista de dicion√°rios com dados dos passageiros

    Returns:
        list: Lista apenas com sobreviventes
    """
    # === SEU C√ìDIGO AQUI ===

    # Usar list comprehension para filtrar
    sobreviventes = [p for p in dados if p['Survived'] == 1]

    return sobreviventes

    # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 4: LIST COMPREHENSIONS ANINHADAS (Bloco 2) ===

## Objetivo
Usar list comprehension aninhada para "achatar" (flatten) uma lista de listas.

## Tarefa
Transformar uma lista de listas em uma lista simples (1D).

Exemplo:
```python
[[1, 2], [3, 4], [5, 6]]  ‚Üí  [1, 2, 3, 4, 5, 6]
```

## Retorno Esperado
Lista simples com todos os elementos das sublistas.

## Dicas
```python
# Comprehension aninhada para flatten
resultado = [item for sublista in lista_de_listas for item in sublista]

# Leia da ESQUERDA para DIREITA:
# "Para cada sublista na lista_de_listas"
# "Para cada item na sublista"
# "Adicione item ao resultado"

# Com filtro (apenas pares)
pares = [item for sublista in lista_de_listas
         for item in sublista
         if item % 2 == 0]
```

## Conceito Importante

- List comprehension aninhada = loop dentro de loop
- Ordem: `for` externo, depois `for` interno
- Muito usado para processar estruturas aninhada

````python
def achatar_dados(lista_aninhada, apenas_pares=False):
    """
    Achata (flatten) uma lista de listas usando comprehension aninhada.

    Args:
        lista_aninhada (list): Lista contendo sublistas
        apenas_pares (bool): Se True, retorna apenas n√∫meros pares

    Returns:
        list: Lista simples (1D) com todos os elementos

    Exemplo:
        >>> achatar_dados([[1, 2], [3, 4]])
        [1, 2, 3, 4]
    """
    # === SEU C√ìDIGO AQUI ===

    if apenas_pares:
        # Achatar E filtrar apenas pares
        resultado = [item for sublista in lista_aninhada
                     for item in sublista
                     if item % 2 == 0]
    else:
        # Achatar tudo
        resultado = [item for sublista in lista_aninhada
                     for item in sublista]

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar a fun√ß√£o

print("\n=== LIST COMPREHENSION ANINHADA: FLATTEN ===\n")

# Teste 1: Flatten simples
print("üìù Teste 1: Achatar lista de listas")
dados_aninhados = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print(f"   Original: {dados_aninhados}")
resultado1 = achatar_dados(dados_aninhados)
print(f"   Achatado: {resultado1}")

# Teste 2: Flatten com dados dos passageiros
print("\nüìù Teste 2: Achatar IDs por classe")
ids_por_classe = [
    [p['PassengerId'] for p in agrupados[0]],  # Classe 1
    [p['PassengerId'] for p in agrupados[1]],  # Classe 2
    [p['PassengerId'] for p in agrupados[2]]   # Classe 3
]
print(f"   Por classe: {ids_por_classe}")
todos_ids = achatar_dados(ids_por_classe)
print(f"   Todos: {todos_ids}")

# Teste 3: Flatten com filtro
print("\nüìù Teste 3: Achatar e filtrar apenas pares")
numeros = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"   Original: {numeros}")
apenas_pares = achatar_dados(numeros, apenas_pares=True)
print(f"   Apenas pares: {apenas_pares}")

print("\nüí° Comprehension aninhada: eficiente e leg√≠vel!")

# Verificar teste
grader.check("p4")
````

## Objetivo
Usar dictionary comprehension para agrupar nomes de passageiros por classe.

## Tarefa
Criar um dicion√°rio onde:
- Chaves: n√∫meros das classes (1, 2, 3)
- Valores: listas com nomes dos passageiros de cada classe

## Retorno Esperado
```python
{
    1: ['Nome1', 'Nome2', ...],
    2: ['Nome3', 'Nome4', ...],
    3: ['Nome5', 'Nome6', ...]
}
```
## Dicas

```python
# Dictionary comprehension b√°sica
resultado = {chave: valor for item in lista}

# Com defaultdict para agrupar
from collections import defaultdict
grupos = defaultdict(list)
for p in dados:
    grupos[p['Pclass']].append(p['Name'])

# Converter defaultdict para dict normal
resultado = dict(grupos)

# Garantir todas as classes (1, 2, 3)
resultado = {classe: grupos.get(classe, []) for classe in [1, 2, 3]}
```

## Conceito Importante

- Dict comprehension cria dicion√°rios de forma concisa
- √ötil para transformar/filtrar/agrupar dados
- Pode combinar com outras estruturas (`defaultdict`, `Counter`)

````python
from collections import defaultdict

def agrupar_nomes_por_classe(dados):
    """
    Agrupa nomes de passageiros por classe usando dict comprehension.

    Args:
        dados (list): Lista de dicion√°rios com dados dos passageiros

    Returns:
        dict: Dicion√°rio com chaves 1, 2, 3 e valores = listas de nomes

    Exemplo:
        >>> agrupar_nomes_por_classe([{'Pclass': 1, 'Name': 'Ana'}])
        {1: ['Ana'], 2: [], 3: []}
    """
    # === SEU C√ìDIGO AQUI ===

    # Agrupar usando defaultdict
    grupos = defaultdict(list)
    for passageiro in dados:
        grupos[passageiro['Pclass']].append(passageiro['Name'])

    # Criar dicion√°rio com todas as classes (1, 2, 3)
    # Usar dict comprehension para garantir que todas existam
    resultado = {classe: grupos.get(classe, []) for classe in [1, 2, 3]}

    return resultado

    # === FIM DO SEU C√ìDIGO ===
````

## Objetivo

## Tarefa
Implementar uma fun√ß√£o generator que:
- Itera sobre lista de passageiros
- Usa `yield` para retornar cada idade
- Pula passageiros com idade None

## Retorno Esperado
Generator que produz idades uma por vez.

## Dicas
```python
# Fun√ß√£o normal retorna lista (tudo de uma vez)
def funcao_normal(dados):
    resultado = []
    for item in dados:
        resultado.append(item)
    return resultado

# Generator usa yield (um de cada vez)
def meu_generator(dados):
    for item in dados:
        yield item  # Produz valor e pausa

# Usar generator
gen = meu_generator([1, 2, 3])
print(next(gen))  # 1
print(next(gen))  # 2
# Ou converter para lista
lista = list(meu_generator([1, 2, 3]))
```

## Conceito Importante

- Generator = iterador que produz valores sob demanda
- `yield` pausa a fun√ß√£o e retorna valor
- Eficiente em mem√≥ria (n√£o cria lista completa)
- Ideal para grandes datasets



````python
def gerar_idades(dados):
    """
    Generator que produz idades dos passageiros.

    Args:
        dados (list): Lista de dicion√°rios com dados dos passageiros

    Yields:
        float: Idade de cada passageiro (pula se None)

    Exemplo:
        >>> dados = [{'Age': 20}, {'Age': 30}]
        >>> list(gerar_idades(dados))
        [20, 30]
    """
    # === SEU C√ìDIGO AQUI ===

    # Iterar sobre passageiros
    for passageiro in dados:
        idade = passageiro.get('Age')

        # Pular se idade for None
        if idade is not None:
            yield idade  # Produz valor e pausa

    # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 7: ITERATOR CUSTOMIZADO (Bloco 2) ===

## Objetivo
Criar uma classe iterator que enumera passageiros (como enumerate, mas customizado).

## Tarefa
Implementar classe `ContadorPassageiros` que:
- Itera sobre lista de passageiros
- Retorna tupla (n√∫mero, nome) para cada um
- Implementa protocolo de itera√ß√£o (__iter__ e __next__)

## Retorno Esperado
Iterator que produz tuplas (1, 'Nome1'), (2, 'Nome2'), ...

## Dicas
```python
# Protocolo de Iterator
class MeuIterator:
    def __init__(self, dados):
        self.dados = dados
        self.index = 0
    
    def __iter__(self):
        return self  # Retorna a si mesmo
    
    def __next__(self):
        if self.index >= len(self.dados):
            raise StopIteration  # Acabaram os elementos
        
        valor = self.dados[self.index]
        self.index += 1
        return valor

# Usar
it = MeuIterator([1, 2, 3])
for valor in it:
    print(valor)
```

## Conceito Importante

- Iterator = objeto que implementa **iter** e **next**
- Generator √© forma simplificada de criar iterator
- **iter**: retorna o iterator (self)
- **next**: retorna pr√≥ximo valor ou StopIteration

````python
class ContadorPassageiros:
    """
    Iterator que enumera passageiros retornando (n√∫mero, nome).

    Similar ao enumerate(), mas customizado para passageiros.

    Exemplo:
        >>> dados = [{'Name': 'Ana'}, {'Name': 'Bruno'}]
        >>> for num, nome in ContadorPassageiros(dados):
        ...     print(f"{num}. {nome}")
        1. Ana
        2. Bruno
    """

    def __init__(self, dados):
        """
        Inicializa o iterator.

        Args:
            dados (list): Lista de dicion√°rios com dados dos passageiros
        """
        # === SEU C√ìDIGO AQUI ===

        self.dados = dados
        self.index = 0  # √çndice atual

        # === FIM DO SEU C√ìDIGO ===

    def __iter__(self):
        """Retorna o pr√≥prio iterator."""
        # === SEU C√ìDIGO AQUI ===

        return self

        # === FIM DO SEU C√ìDIGO ===

    def __next__(self):
        """
        Retorna pr√≥ximo item ou lan√ßa StopIteration.

        Returns:
            tuple: (n√∫mero, nome) do passageiro
        """
        # === SEU C√ìDIGO AQUI ===

        # Verificar se ainda h√° elementos
        if self.index >= len(self.dados):
            raise StopIteration

        # Pegar passageiro atual
        passageiro = self.dados[self.index]
        numero = self.index + 1  # Come√ßar de 1, n√£o 0
        nome = passageiro['Name']

        # Avan√ßar √≠ndice
        self.index += 1

        # Retornar tupla (n√∫mero, nome)
        return (numero, nome)

        # === FIM DO SEU C√ìDIGO ===
````

# === PR√ÅTICA 9: MEDI√á√ÉO DE PERFORMANCE (Bloco 3) ===

## Objetivo
Usar `timeit` para comparar a performance de busca em lista vs set.

## Tarefa
Comparar o tempo de busca de um elemento em:
- Lista (busca linear - O(n))
- Set (busca com hash - O(1))

## Retorno Esperado
Dicion√°rio com:
```python
{
    'tempo_lista': float,    # Tempo em segundos
    'tempo_set': float,      # Tempo em segundos
    'diferenca': float       # Fator de diferen√ßa (tempo_lista / tempo_set)
}
```
## Dicas
```
import timeit

# Medir tempo de uma opera√ß√£o
tempo = timeit.timeit(
    stmt='2 in lista',           # C√≥digo a executar
    setup='lista = [1, 2, 3]',   # Setup antes de executar
    number=10000                  # Quantas vezes executar
)

# Para usar vari√°veis existentes, use globals()
tempo = timeit.timeit(
    stmt='valor in lista',
    globals={'lista': minha_lista, 'valor': meu_valor},
    number=10000
)

# Calcular diferen√ßa
diferenca = tempo_maior / tempo_menor
```

## Conceitos Importantes

- Lista: busca O(n) - precisa percorrer elementos
- Set: busca O(1) - acesso direto por hash
- Quanto maior o dataset, maior a diferen√ßa!

````python
import timeit

def comparar_busca(dados, valor):
    """
    Compara performance de busca em lista vs set usando timeit.

    Args:
        dados (list): Lista de valores para buscar
        valor: Valor a procurar nos dados

    Returns:
        dict: Dicion√°rio com 'tempo_lista', 'tempo_set', 'diferenca'

    Exemplo:
        >>> comparar_busca([1, 2, 3, 4, 5], 3)
        {'tempo_lista': 0.00012, 'tempo_set': 0.00008, 'diferenca': 1.5}
    """
    # === SEU C√ìDIGO AQUI ===

    # Converter dados para set uma √∫nica vez (n√£o contar no tempo)
    dados_set = set(dados)

    # Medir tempo de busca na lista
    # number=10000 significa executar 10000 vezes para ter medida precisa
    tempo_lista = timeit.timeit(
        stmt='valor in dados_lista',
        globals={'dados_lista': dados, 'valor': valor},
        number=10000
    )

    # Medir tempo de busca no set
    tempo_set = timeit.timeit(
        stmt='valor in dados_set',
        globals={'dados_set': dados_set, 'valor': valor},
        number=10000
    )

    # Calcular diferen√ßa (quantas vezes lista √© mais lenta)
    diferenca = tempo_lista / tempo_set if tempo_set > 0 else 0

    return {
        'tempo_lista': tempo_lista,
        'tempo_set': tempo_set,
        'diferenca': diferenca
    }

    # === FIM DO SEU C√ìDIGO ===
````

````python
# Testar a fun√ß√£o com diferentes tamanhos de dataset

print("\n=== COMPARA√á√ÉO DE PERFORMANCE: LISTA vs SET ===\n")

# Teste 1: Dataset pequeno (8 passageiros)
print("üìä Teste 1: Dataset pequeno (8 elementos)")
ids_pequeno = [p['PassengerId'] for p in passageiros]
resultado1 = comparar_busca(ids_pequeno, 5)

print(f"   Tempo lista: {resultado1['tempo_lista']:.6f}s")
print(f"   Tempo set:   {resultado1['tempo_set']:.6f}s")
print(f"   Diferen√ßa:   {resultado1['diferenca']:.2f}x mais r√°pido")

# Teste 2: Dataset m√©dio (100 elementos)
print("\nüìä Teste 2: Dataset m√©dio (100 elementos)")
dados_medio = list(range(100))
resultado2 = comparar_busca(dados_medio, 99)  # Buscar √∫ltimo elemento

print(f"   Tempo lista: {resultado2['tempo_lista']:.6f}s")
print(f"   Tempo set:   {resultado2['tempo_set']:.6f}s")
print(f"   Diferen√ßa:   {resultado2['diferenca']:.2f}x mais r√°pido")

# Teste 3: Dataset grande (10.000 elementos)
print("\nüìä Teste 3: Dataset grande (10.000 elementos)")
dados_grande = list(range(10000))
resultado3 = comparar_busca(dados_grande, 9999)  # Buscar √∫ltimo elemento

print(f"   Tempo lista: {resultado3['tempo_lista']:.6f}s")
print(f"   Tempo set:   {resultado3['tempo_set']:.6f}s")
print(f"   Diferen√ßa:   {resultado3['diferenca']:.2f}x mais r√°pido")

# An√°lise
print("\n" + "="*60)
print("üìà AN√ÅLISE:")
print("="*60)
print(f"Com 8 elementos:     Set √© {resultado1['diferenca']:.1f}x mais r√°pido")
print(f"Com 100 elementos:   Set √© {resultado2['diferenca']:.1f}x mais r√°pido")
print(f"Com 10.000 elementos: Set √© {resultado3['diferenca']:.1f}x mais r√°pido")
print("\nüí° CONCLUS√ÉO:")
print("Quanto maior o dataset, maior a vantagem do set!")
print("Lista: O(n) - tempo cresce linearmente")
print("Set: O(1) - tempo constante")
print("="*60)

# Verificar teste
grader.check("p9")
````
