# === PRÁTICA 1: ESTATÍSTICAS DESCRITIVAS ===

## Objetivo
Calcular as principais estatísticas descritivas.

## Tarefa
Dado DataFrame com coluna 'valor', retorne dicionário com:
- media: média dos valores
- mediana: mediana dos valores
- desvio_padrao: desvio padrão
- minimo: valor mínimo
- maximo: valor máximo

## Retorno Esperado
Dicionário com as 5 estatísticas.

## Dicas
```python
# Estatísticas Pandas
df['coluna'].mean()    # Média
df['coluna'].median()  # Mediana
df['coluna'].std()     # Desvio padrão
df['coluna'].min()     # Mínimo
df['coluna'].max()     # Máximo
```

**Tempo estimado:** 5 minutos

````python
def calcular_estatisticas_basicas(df):
    """
    Calcula estatísticas descritivas da coluna 'valor'.
    
    Args:
        df (pd.DataFrame): DataFrame com coluna 'valor'
        
    Returns:
        dict: Dicionário com estatísticas
    """
    # === SEU CÓDIGO AQUI ===
    
    stats = {
        'media': df['valor'].mean(),
        'mediana': df['valor'].median(),
        'desvio_padrao': df['valor'].std(),
        'minimo': df['valor'].min(),
        'maximo': df['valor'].max()
    }
    
    return stats
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar DataFrame de notas
np.random.seed(42)
df_notas = pd.DataFrame({
    'aluno': [f'Aluno {i+1}' for i in range(20)],
    'valor': np.random.uniform(5, 10, 20)
})

print("\n=== NOTAS DOS ALUNOS ===")
print(df_notas.head(10))
print(f"\n... (total de {len(df_notas)} alunos)")

# Calcular estatísticas
stats = calcular_estatisticas_basicas(df_notas)

print("\n=== ESTATÍSTICAS DESCRITIVAS ===")
for chave, valor in stats.items():
    print(f"{chave.replace('_', ' ').title()}: {valor:.2f}")

# Comparar com describe()
print("\n=== COMPARAÇÃO COM describe() ===")
print(df_notas['valor'].describe())

# Verificar teste
grader.check("p1")
````

# === PRÁTICA 2: QUARTIS ===

## Objetivo
Calcular os quartis (Q1, Q2, Q3).

## Tarefa
Dado DataFrame com 'valor', retorne tupla com (Q1, Q2, Q3).
- Q1: 25º percentil
- Q2: 50º percentil (mediana)
- Q3: 75º percentil

## Retorno Esperado
Tupla (Q1, Q2, Q3).

## Dicas
```python
# Calcular quartis
Q1 = df['col'].quantile(0.25)
Q2 = df['col'].quantile(0.50)  # ou .median()
Q3 = df['col'].quantile(0.75)
```

**Tempo estimado:** 5 minutos

````python
def calcular_quartis(df):
    """
    Calcula quartis da coluna 'valor'.
    
    Args:
        df (pd.DataFrame): DataFrame com coluna 'valor'
        
    Returns:
        tuple: (Q1, Q2, Q3)
    """
    # === SEU CÓDIGO AQUI ===
    
    Q1 = df['valor'].quantile(0.25)
    Q2 = df['valor'].quantile(0.50)
    Q3 = df['valor'].quantile(0.75)
    
    return (Q1, Q2, Q3)
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Usar dados de notas
print("\n=== DISTRIBUIÇÃO DAS NOTAS ===")
print(df_notas['valor'].describe())

# Calcular quartis
Q1, Q2, Q3 = calcular_quartis(df_notas)

print("\n=== QUARTIS ===")
print(f"Q1 (25%): {Q1:.2f}")
print(f"Q2 (50% - Mediana): {Q2:.2f}")
print(f"Q3 (75%): {Q3:.2f}")

# Calcular IQR
IQR = Q3 - Q1
print(f"\nIQR (Amplitude Interquartil): {IQR:.2f}")

# Interpretar
print("\n=== INTERPRETAÇÃO ===")
print(f"25% dos alunos tiraram nota até {Q1:.2f}")
print(f"50% dos alunos tiraram nota até {Q2:.2f} (mediana)")
print(f"75% dos alunos tiraram nota até {Q3:.2f}")
print(f"\n50% centrais das notas variam em {IQR:.2f} pontos")

# Verificar teste
grader.check("p2")
````

# === PRÁTICA 3: IDENTIFICAR OUTLIERS ===

## Objetivo
Identificar outliers usando regra IQR.

## Tarefa
Dado DataFrame com 'valor', retorne Series com valores outliers.

Regra: outlier se:
- valor < Q1 - 1.5 × IQR
- valor > Q3 + 1.5 × IQR

## Retorno Esperado
Series com apenas os outliers.

## Dicas
```python
# Calcular limites
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar outliers
outliers = df[(df['val'] < lim_inf) | (df['val'] > lim_sup)]['val']
```

**Tempo estimado:** 7 minutos

````python
def identificar_outliers_iqr(df):
    """
    Identifica outliers usando regra IQR.
    
    Args:
        df (pd.DataFrame): DataFrame com coluna 'valor'
        
    Returns:
        pd.Series: Series com outliers
    """
    # === SEU CÓDIGO AQUI ===
    
    # Calcular quartis
    Q1 = df['valor'].quantile(0.25)
    Q3 = df['valor'].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calcular limites
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Filtrar outliers
    outliers = df[(df['valor'] < limite_inferior) | (df['valor'] > limite_superior)]['valor']
    
    return outliers
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar dados com outliers
np.random.seed(42)
valores_normais = np.random.normal(100, 15, 50)
valores_outliers = [30, 180, 190]  # Outliers artificiais

df_com_outliers = pd.DataFrame({
    'valor': np.concatenate([valores_normais, valores_outliers])
})

print("\n=== DADOS COM OUTLIERS ===")
print(df_com_outliers['valor'].describe())

# Identificar outliers
outliers = identificar_outliers_iqr(df_com_outliers)

print(f"\n=== OUTLIERS IDENTIFICADOS ===")
print(f"Quantidade: {len(outliers)}")
print(f"Valores: {sorted(outliers.values)}")

# Calcular limites para mostrar
Q1 = df_com_outliers['valor'].quantile(0.25)
Q3 = df_com_outliers['valor'].quantile(0.75)
IQR = Q3 - Q1
lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

print(f"\n=== LIMITES ===")
print(f"Limite inferior: {lim_inf:.2f}")
print(f"Limite superior: {lim_sup:.2f}")
print(f"\nQualquer valor fora deste intervalo é outlier!")

# Verificar teste
grader.check("p3")
````

# === PRÁTICA 4: HISTOGRAMA ===

## Objetivo
Criar histograma de uma variável.

## Tarefa
Dado DataFrame com 'valor', crie histograma com:
- 20 bins
- Cor azul ('steelblue')
- Borda preta
- Título, labels nos eixos

## Retorno Esperado
Tupla (fig, ax) com o gráfico.

## Dicas
```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['col'], bins=20, color='steelblue', edgecolor='black')
ax.set_title('Título')
ax.set_xlabel('Label X')
ax.set_ylabel('Label Y')
```

**Tempo estimado:** 6 minutos

````python
def criar_histograma(df):
    """
    Cria histograma da coluna 'valor'.
    
    Args:
        df (pd.DataFrame): DataFrame com coluna 'valor'
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU CÓDIGO AQUI ===
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Criar histograma
    ax.hist(df['valor'], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
    
    # Adicionar labels
    ax.set_title('Distribuição de Valores', fontsize=14, fontweight='bold')
    ax.set_xlabel('Valor', fontsize=12)
    ax.set_ylabel('Frequência', fontsize=12)
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar dados normalmente distribuídos
np.random.seed(42)
df_dist = pd.DataFrame({
    'valor': np.random.normal(100, 15, 200)
})

print("\n=== ESTATÍSTICAS DOS DADOS ===")
print(df_dist['valor'].describe())

# Criar histograma
fig, ax = criar_histograma(df_dist)
plt.show()

print("\n=== INTERPRETAÇÃO ===")
print("O histograma mostra a forma da distribuição:")
print("- Centro: onde a maioria dos valores está")
print("- Largura: o quão dispersos são os valores")
print("- Forma: simétrica, assimétrica, bimodal, etc.")

# Verificar teste
grader.check("p4")
````

# === PRÁTICA 5: BOXPLOT POR CATEGORIA ===

## Objetivo
Criar boxplot comparando categorias.

## Tarefa
Dado DataFrame com 'categoria' e 'valor',
crie boxplot mostrando distribuição de valores por categoria.

## Retorno Esperado
Tupla (fig, ax) com o gráfico.

## Dicas
```python
# Preparar dados por categoria
data_by_cat = [df[df['cat']==c]['val'] for c in df['cat'].unique()]

# Criar boxplot
ax.boxplot(data_by_cat, labels=df['cat'].unique())
```

**Tempo estimado:** 7 minutos

````python
def criar_boxplot_por_categoria(df):
    """
    Cria boxplot de 'valor' por 'categoria'.
    
    Args:
        df (pd.DataFrame): DataFrame com 'categoria' e 'valor'
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU CÓDIGO AQUI ===
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Preparar dados por categoria
    categorias = df['categoria'].unique()
    data_by_cat = [df[df['categoria'] == cat]['valor'].values for cat in categorias]
    
    # Criar boxplot
    ax.boxplot(data_by_cat, labels=categorias)
    
    # Adicionar labels
    ax.set_title('Boxplot de Valores por Categoria', fontsize=14, fontweight='bold')
    ax.set_xlabel('Categoria', fontsize=12)
    ax.set_ylabel('Valor', fontsize=12)
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Criar dados com 3 categorias
np.random.seed(42)
df_categorias = pd.DataFrame({
    'categoria': ['A']*50 + ['B']*50 + ['C']*50,
    'valor': np.concatenate([
        np.random.normal(80, 10, 50),   # Categoria A
        np.random.normal(100, 15, 50),  # Categoria B
        np.random.normal(120, 12, 50)   # Categoria C
    ])
})

print("\n=== ESTATÍSTICAS POR CATEGORIA ===")
print(df_categorias.groupby('categoria')['valor'].describe())

# Criar boxplot
fig, ax = criar_boxplot_por_categoria(df_categorias)
plt.show()

print("\n=== INTERPRETAÇÃO DO BOXPLOT ===")
print("- Caixa: 50% centrais (Q1 a Q3)")
print("- Linha na caixa: mediana (Q2)")
print("- Whiskers (bigodes): valores dentro do range normal")
print("- Pontos isolados: outliers")

# Verificar teste
grader.check("p5")
````

# === PRÁTICA 6: BAR CHART ===

## Objetivo
Criar bar chart de contagens ou agregações.

## Tarefa
Dado DataFrame com 'categoria' e 'valor',
crie bar chart mostrando média de valores por categoria.

## Retorno Esperado
Tupla (fig, ax) com o gráfico.

## Dicas
```python
# Calcular médias por categoria
medias = df.groupby('categoria')['valor'].mean()

# Criar bar chart
ax.bar(medias.index, medias.values)
```

**Tempo estimado:** 6 minutos

````python
def criar_bar_chart(df):
    """
    Cria bar chart de média por categoria.
    
    Args:
        df (pd.DataFrame): DataFrame com 'categoria' e 'valor'
        
    Returns:
        tuple: (fig, ax)
    """
    # === SEU CÓDIGO AQUI ===
    
    # Calcular médias
    medias = df.groupby('categoria')['valor'].mean()
    
    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Criar bar chart
    ax.bar(medias.index, medias.values, color='coral', edgecolor='black', alpha=0.7)
    
    # Adicionar labels
    ax.set_title('Média de Valores por Categoria', fontsize=14, fontweight='bold')
    ax.set_xlabel('Categoria', fontsize=12)
    ax.set_ylabel('Valor Médio', fontsize=12)
    ax.grid(True, alpha=0.3, axis='y')
    
    # Adicionar valores nas barras
    for i, (cat, media) in enumerate(medias.items()):
        ax.text(i, media + 2, f'{media:.1f}', ha='center', fontweight='bold')
    
    plt.tight_layout()
    
    return fig, ax
    
    # === FIM DO SEU CÓDIGO ===
````

````python
# Usar dados das categorias
print("\n=== MÉDIAS POR CATEGORIA ===")
medias_por_cat = df_categorias.groupby('categoria')['valor'].mean()
print(medias_por_cat)

# Criar bar chart
fig, ax = criar_bar_chart(df_categorias)
plt.show()

print("\n=== ANÁLISE ===")
print(f"Categoria com maior média: {medias_por_cat.idxmax()}")
print(f"Valor: {medias_por_cat.max():.2f}")
print(f"\nCategoria com menor média: {medias_por_cat.idxmin()}")
print(f"Valor: {medias_por_cat.min():.2f}")

# Verificar teste
grader.check("p6")
````
