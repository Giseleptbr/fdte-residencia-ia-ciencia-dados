# 29/10 - Desempenho e Avaliação II

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.metrics import r2_score, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# ABALON
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
cols = ['Sex','Length','Diameter','Height','Whole_weight','Shucked_weight','Viscera_weight','Shell_weight','Rings']
data = pd.read_csv(url, names=cols)

data = pd.get_dummies(data, columns=['Sex'], drop_first=True)

# rings
X = data[['Length','Diameter','Whole_weight']]
y = data[['Rings']]
```

Regressão

```python
r2_train = []
r2_test = []

for i in np.arange(5, X.shape[0]-5, 50):
    X_train = X.loc[:i, :]
    y_train = y.loc[:i, :]
    X_test = X.loc[i+1:, :]
    y_test = y.loc[i+1:, :]

    reg = DecisionTreeRegressor()
    reg.fit(X_train, y_train)

    y_train_pred = reg.predict(X_train)
    y_test_pred = reg.predict(X_test)

    e_train = r2_score(y_train, y_train_pred)
    e_test = r2_score(y_test, y_test_pred)

    r2_train.append(e_train)
    r2_test.append(e_test)
    print(f"R²_teste={e_test:.3f}, R²_treino={e_train:.3f}")
```

Classificação (transformando variável alvo)

```python
# Variável alvo categórica (0 = abaixo da média, 1 = acima)
media_aneis = y['Rings'].mean()
y_class = (y['Rings'] > media_aneis).astype(int)

acc_train = []
acc_test = []

for i in np.arange(5, X.shape[0]-5, 50):
    X_train = X.loc[:i, :]
    y_train = y_class.loc[:i]
    X_test = X.loc[i+1:, :]
    y_test = y_class.loc[i+1:]

    clf = DecisionTreeClassifier()
    clf.fit(X_train, y_train)

    y_train_pred = clf.predict(X_train)
    y_test_pred = clf.predict(X_test)

    e_train = accuracy_score(y_train, y_train_pred)
    e_test = accuracy_score(y_test, y_test_pred)

    acc_train.append(e_train)
    acc_test.append(e_test)
    print(f"Acurácia_teste={e_test:.3f}, Acurácia_treino={e_train:.3f}")
```

Mudando a variável

```python
X = data[['Length', 'Diameter', 'Whole_weight']]
y = data[['Shucked_weight']]  # nova variável alvo (em vez de Rings)

r2_train = []
r2_test = []

for i in np.arange(5, X.shape[0]-5, 50):
    X_train = X.loc[:i, :]
    y_train = y.loc[:i, :]
    X_test = X.loc[i+1:, :]
    y_test = y.loc[i+1:, :]

    reg = DecisionTreeRegressor()
    reg.fit(X_train, y_train)

    y_train_pred = reg.predict(X_train)
    y_test_pred = reg.predict(X_test)

    e_train = r2_score(y_train, y_train_pred)
    e_test = r2_score(y_test, y_test_pred)

    r2_train.append(e_train)
    r2_test.append(e_test)
    print(f"R²_teste={e_test:.3f}, R²_treino={e_train:.3f}")
```
