# Ex 4
- Mostre o maior erro da diferença de pesos das partes com o peso total.

- Selecione as linhas em que este erro é no máximo 10% do maior erro.

```python
X['erro'] = X['Shucked_weight']+X['Viscera_weight']+X['Shell_weight']-X['Whole_weight']
X['erro'] = X['erro'].abs()
```

```python
X['erro'].max()
```

```python
X.loc[X['erro']<=0.1*X['erro'].max(),:]
```
